# 第九章 相关性回归

回归分析（Regression Analysis）是一种分析数据的统计学方法，是估计两个或多个变量间数量关系的数学模型，并能用于预测研究者感兴趣的变量。例如：研究收入与受教育程度之间的数量关系。线性回归是回归分析的一种特例，它用于分析因变量为数值变量，自变量为数值变量或分类变量之间的关系。例如：
* GDP与人口增长率、进出口贸易额、投资等的关系
* 股票价格波动率与上市公司盈利能力、市场环境、无风险利率等的关系
* 人体健康指数与性别、年龄、职业等的关系
* 犯罪率与人均GDP、CPI、治安水平、失业了等的关系
一元线性回归用于研究一个自变量与一个因变量的**数量关系**；或利用一个自变量预测一个因变量的**发展趋势**；
例：人均收入$X$ 与人均食品消费的关系

## 9.1 相关性度量

### 9.1.1 相关系数


相关系数(correlation coefficient)是根据样本数据计算的**度量两个变量之间线性关系强度**的统计量。若相关系数是根据总体全部数据计算的，称为总体相关系数，记为$ \rho $;   若是根据样本数据计算的，则称为样本相关系数，记为$r$。样本相关系数的计算公式为:
$$
r=\frac{\sum\left(x_i-\bar{x}\right)\left(y_i-\bar{y}\right)}{\sqrt{\sum\left(x_i-\bar{x}\right)^2 \sum\left(y_i-\bar{y}\right)^2}}
$$
按照该公式定义的相关系数也称为 Pearson 相关系数(Pearson’s correlation coefficient)

**相关系数的性质可总结如下：**<p>
（1）$r$的取值范围是$[-1，1]$，即$一1\leq{r}\leq 1$。若$0< r\leq 1$，表明$x$与$y$之间存在正线性相关关系;若$-1\leq r<0$，表明$x$与$y$之间存在负线性相关关系；若$r=+1$，表明$x$与$y$之间为完全正线性相关关系;若$r=-1$，表明$x$与$y$之间为完全负线性相关关系。可见当$|r|=1$时，$y$的取值完全依赖于$x$，二者之间即为函数关系;当 $r$=0时，说明$y$的取值与$x$无关，即二者之间不存在线性相关关系。<br>
（2）$r$具有对称性。$x$与$y$之间的相关系数$r_{xy}$和$y$与$x$之间的相关系数$r_{yx}$相等，
即$r_{xy}=r_{yx}$。<br>
（3）$r$的数值大小与$x$和$y$的原点及尺度无关。改变$x$和$y$的数据原点及计量尺度，并不改变$r$的数值大小。<br>
（4）$r$仅仅是$x$与$y$之间线性关系的一个度量，它不能用于描述非线性关系。这意味着，$r$=0 只表示两个变量之间不存在线性相关关系，并不说明变量之间没有任何关系，它们之间可能存在非线性相关关系。变量之间的非线性相关程度较大时，就可能会导致$r$=0。因此当$r=0$或很小时，不能轻易得出两个变量之间不存在相关关系的结论，而应该结合下文的散点图作出合理的解释。<br>
（5）$r$虽然是两个变量之间线性关系的一个度量，却不意味着$x$与$y$一定有因果关系。了解相关系数的性质有助于对其实际意义作出解释。但根据实际数据计算出的$r$，其取值一般在$-1\sim 1$之间，$|r|→ 1$说明两个变量之间的线性关系越强；$|r|→ 0$说明两个变量之间的线性关系越弱。<br>
（此部分来源于统计学 贾俊平 第六版 P269）

### 9.1.2 散点图

散点图：对于两个变量$x$和$y$，通过观察或试验可以得到若干组数据，记为($x_i$，$y_i$)=($i$=1，2，…，n)。用坐标的横轴代表变量$x$，纵轴代表变量$y$，每组数据($x_i$，$y_i$) 在坐标系中用一个点表示，n组数据在坐标系中形成的n个点称为散点，由坐标及其散点形成的二维数据图称为散点图(scatter diagram)。<br>
**散点图**是描述变量之间关系的一种直观方法，从中可以大体上看出变量之间的关系形态及关系强度。可以通过绘制散点图来判断变量之间的关系形态，如果是线性关系，则可以利用相关系数来测度两个变量之间的关系强度，然后对相关系数进行显著性检验，以判断样本所反映的关系能否代表两个变量总体上的关系。从下面的散点图可以看出，相关关系的表现形态大体上可分为线性相关、非线性相关、完全相关和不相关等几种。就两个变量而言，
* 如果变量之间的关系近似地表现为一条直线，则称为线性相关;
* 如果变量之间的关系近似地表现为一条曲线，则称为非线性相关或曲线相关;
* 如果一个变量的取值完全依赖于另一个变量，各观测点落在一条直线上，则称为完全相关;
* 如果两个变量的观测点很分散，无任何规律，则表示变量之间没有相关关系。
* 在线性相关中，若两个变量的变动方向相同，一个变量的数值增加，另一个变量的数值也随之增加，或一个变量的数值减少，另一个变量的数值也随之减少，则称为正相关;
* 若两个变量的变动方向相反，一个变量的数值增加，另一个变量的数值随之减少，或一个变量的数值减少，另一个变量的数值随之增加，则称为负相关。
（此部分来源于统计学 贾俊平 第六版 P266）


* 第一幅图：随着$x$的增长，$y$也会增长，$x$和$y$呈现正相关关系
* 第二幅图：随着$x$的增长，$y$也会降低，$x$和$y$呈现负相关关系
* 第三幅和第四幅图：散点在同一条直线上，说明$x$与$y$存在完全相关关系
* 第五幅图：随着$x$的增加，$y$先增长再下降，呈现类似抛物线的趋势，说明它们存在非线性相关关系
* 最后一幅图：随着$x$的增加，$y$有时候增长，有时候下降，说明它们关系不确定的，因此这两个变量无关

### 9.1.3 相关性检验

一般情况下，总体相关系数$\rho$是未知的，通常是将样本相关系数$r$作为$\rho$的近似估计值。但由于$r$是根据样本数据计算出来的，因此会受到抽样波动的影响。由于取的样本不同，$r$的取值也就不同，因此$r$是一个随机变量。能否根据样本相关系数说明总体的相关程度呢?这就需要考察样本相关系数的可靠性，也就是进行显著性检验。<br>
**（1）$r$的抽样分布：**<br>
为了对样本相关系数r的显著性进行检验，需要考察$r$的抽样分布。$r$的抽样分布随总体相关系数$\rho$和样本量n的大小而变化。当样本数据来自正态总体时，随着$n$的增大，$r$的抽样分布趋于正态分布，尤其是在总体相关系数$\rho$很小或接近0时，趋于正态分布的趋势非常明显。而当$\rho$远离0时，除非n非常大，否则$r$的抽样分布呈现一定的偏态。因为$r$是在$\rho$的周围分布的，当$\rho$的数值接近+1或-1时，比如$\rho$=0.96，$r$的值可能以 0.96 为中心向两个方向变化，又由于$r$的取值范围在$-1\sim +1$之间，所以一方的变化以+1为限，全距是0.04，而另一方的变化以一1为限，全距是1.96，两个方向变化的全距不等，因此$r$的抽样分布也不可能对称。但当$\rho$等于0或接近0时，两个方向变化的全距接近相等，所以r的抽样分布也就接近对称。<br>
总之，当$\rho$为较大的正值时，$r$呈现左偏分布;当$\rho$为较大的负值时，$r$呈现右偏分布。只有当$\rho$接近0，而样本量n很大时，才能认为$r$是接近正态分布的随机变量。然而，以样本$r$来估计总体$\rho$时，总是假设$r$为正态分布，但这一假设常常会带来一些严重后果。
（此部分来源于统计学 贾俊平 第六版 P270）


**（2）相关系数检验：**<br>相关系数的显著检验通常采用 R.A.Fisher 提出的 t 检验方法。该检验主要用于小样本，也可以用于大样本。它的本质是对总体相关系数 $\rho$ 进行检验，$\rho$  通常是未知的，需要根据样本的相关系数 r 来估计。<br>
相关系数检验过程：<br>
第 1 步: 提出假设<br>
$H_0:\rho = 0$ 或两个变量的线性关系不显著<br>
$H_1:\rho \neq 0$ 或两个变量的线性关系显著<br>
第 2 步: 计算检验统计量<br>
首先，计算样本的相关系数 r，用来估计总体相关系数 $\rho$ 其次，构造 $t$ 统计量，计算公式如下：<br>
$$t=\frac{r\sqrt{n-1}}{1-r^2}\sim t(n-2)$$   <br>
第 3 步：进行决策
给定显著水平 $\alpha$，求出统计量 $t$的 P-value 值，如果P-value<$\alpha$，则拒绝$H_0$, 表示总体的两个变量之间线性关系显著。否则，接受$H_0$, 表示总体的两个变量之间线性关系不显著

## 9.2 一元线性回归模型

由于自变量不能解释$Y$的不确定性（随机性），因此回归模型定义为：
$Y = \beta_0 + X\beta_1+ \epsilon $ 

$\beta_0$ 和 $\beta_1$ 是未知的待估计参数
$\epsilon$ 是误差项，也称为随机干扰项，它反映了 $Y$ 的不确定性，与 $X$ 无关<br>
$\epsilon$满足：(1)$\epsilon$服从正态分布，(2) $E(\epsilon)=0$；(3)$Var(\epsilon_i)= \sigma^2$且$cov(\epsilon_i,\epsilon_j)=0$<br>
以人均收入 $X$ 与人均食品消费的线性模型为例:<br>
$Y$是人均食品消费支出，称为：被解释变量、或因变量；<br>
$X$是人均收入，称为：解释变量、或自变量；<br>
$\beta_0, \beta_1$是未知的待估计参数；<br>
$\epsilon$是随机干扰项， 它反映了 $Y$ 的不确定性，与 $X$ 无关;

如果随机干扰项 $\epsilon$ 的的期望为 0，即满足 E($\epsilon$) = 0。对回归模型取条件期望，则有$$E(Y|X) = \beta_0 + X\beta_1 $$
它表示从“平均”的角度，反映$X$ 与 $Y$ 的关系，在给定参数和样本的情况下，$X$ 与 $Y$ 的关系是确定的线性关系，该模型也称为线性回归方程。

从回归模型的结构可以看出，回归分析模型由**确定性部分**和**不确定部分组成**
 

假设 $\beta_0$ 和 $\beta_1$ 的估计值为 $\hat{\beta_0}$,$\hat{\beta_1}$, 那么 Y 的预测值为 $\hat{Y}$ 。因此， $\hat{Y}$ 是关于 $\hat{\beta_0}$,$\hat{\beta_1}$
的函数，即：$$\hat{Y}=\hat{\beta_0}+X\hat{\beta_1}$$


在该方程中，$\hat{\beta_0}$和$\hat{\beta_1}$是未知的，那么如何求解该参数呢？

### 9.2.1 最小二乘估计

为了由样本数据得到回归参数 $\beta_0$ 和 $\beta_1$ 的理想估计值，我们可以使用最小二乘估计 (ordinary least square estimation,OLSE)。对于每个样本观测值 ($Y_i$, $X_i$)，最小二乘法考虑观测值 $Y_i$ 与其回归值 $E(Y_i) = \beta_0 + X_i\beta_1$ 的离差越小越好，因此定义离差平方和：$$Q=\sum_{i=1}^n(Y_{i}-E(Y_i)^2)=\sum_{i=1}^n(Y_{i}-\beta_0-X_i\beta_1)^2$$

所谓最小二乘法就是求解 $\beta_0$ 和 $\beta_1$ 的估计值 $\hat{\beta_0}$和$\hat{\beta_1}$，使得定义的离差平方和达到最小，即寻找$\hat{\beta_0}$和$\hat{\beta_1}$满足
$$Q(\hat{\beta_0},\hat{\beta_1})=\sum_{i=1}^n(Y_{i}-\beta_0-X_i\beta_1)^2=min_(\hat{\beta_0},\hat{\beta_1})\sum_{i=1}^n(Y_{i}-\beta_0-X_i\beta_1)^2$$

为了求解 $\beta_0$ 和 $\beta_1$的估计值，可以分别求 $\hat{\beta_0}$和$\hat{\beta_1}$ 的偏导数，并使其偏导数为 0，此时构造估计方程组:
$$\begin{gather}
\frac{\partial Q}{\beta_0}|_{\beta_0=\hat\beta_0} = -2\sum_{i=1}^n(Y_i-\hat\beta_0-X_i\hat\beta_1) = 0 \\
\frac{\partial Q}{\beta_1}|_{\beta_1=\hat\beta_1}  = -2\sum_{i=1}^n(Y_i-\hat\beta_0-X_i\hat\beta_1)X_i=0
\end{gather}$$


解方程组，可以得到$\beta_0和\beta_1的估计\hat{\beta_0}和\hat{\beta_1}$
：
\begin{align*}
\hat\beta_0 &= \bar{Y} - \bar{X} \hat\beta_1  \\
\hat\beta_1 &= \frac{\sum_{i=1}^n(X_i-\bar{X})(Y_i-\bar{Y})}{\sum_{i=1}^n(X_i-\bar{X})^2}
\end{align*}



该估计方法称为**最小二乘法**，解得的参数称为**最小二乘的解**。

从该结果可以看出：
* 回归线穿过点$(\bar{X} ,\bar{Y})$
* 若对$X$和$Y$进行标准化，得到标准化的向量$X^\prime$和$Y^\prime$，那么用标准化的数据建立回归模型，其回归系数的估计为$\beta_0=0$和$\beta_1 = \frac{1}{n}(X^\prime)^TY^\prime$，其中$(X^\prime)^TY^\prime$是$X^\prime$与$Y^\prime$的相关系数的样本估计值。


注意：假设自变量$X$的样本方差为$\sigma^2_X=\frac{1}{n}\sum_{i=1}^n(X_i-\bar{X})^2$，那么对样本$X_i$标准化后$X_i^\prime=\frac{X_i-\bar{X}}{\sigma_X}$，同理假设$Y^\prime$是$Y$标准化后的向量，其方差为$\sigma^2_Y$。那么用$X^\prime$与$Y^\prime$建立回归模型，回归系数估计为：

\begin{equation}\label{key}
\beta = \frac{\sum_{i=1}^n(X_i^\prime-\bar{X^\prime})(Y_i^\prime-\bar{Y^\prime}) }{\sum_{i=1}^n(X_i^\prime-\bar{X^\prime})^2}
\end{equation}

其中：$\bar{X^\prime}=\frac{1}{n}\sum_{i=1}^n\frac{X_i-\bar{X_i}}{\sigma_X}=0$， $\bar{Y^\prime}=\frac{1}{n}\sum_{i=1}^n\frac{Y_i-\bar{Y_i}}{\sigma_Y}=0$	
同理，
$$\sum_{i=1}^n(X_i^\prime-\bar{X^\prime})^2=\sum_{i=1}^n(X_i^\prime)^2=\sum_{i=1}^n(\frac{X_i-\bar{X}}{\sigma_X})^2=\frac{\sum_{i=1}^n(X_i-\bar{X})^2}{\sigma_X^2}=\frac{n\sigma_X^2}{\sigma_X^2}=n$$


因此，有\begin{equation}
\beta_1 = \frac{\sum_{i=1}^n X_i^\prime Y_i^\prime}{\sum_{i=1}^n(X_i^\prime)^2}=\frac{1}{n}\sum_{i=1}^n X_i^\prime Y_i^\prime
\end{equation}
因$\bar{X^\prime}=0$、$\bar{Y^\prime}=0$，那么
$$\bar{Y^\prime} = \beta_0 + \bar{X^\prime}\beta_1$$
所以，$\beta_0=0$


因为$X_i^\prime$与$Y_i^\prime$分别是$X_i$和$Y_i$标准化后的值，那么它们的方差分别为：
\begin{equation}
var(X_i^\prime)=var(\frac{X_i-\bar{X}}{\sigma_X})=\frac{1}{\sigma_X^2}var(X_i)=\frac{\sigma_X^2}{\sigma_X^2}=1
\end{equation}
\begin{equation}
var(Y_i^\prime)=var(\frac{Y_i-\bar{Y}}{\sigma_Y})=\frac{1}{\sigma_Y^2}var(Y_i)=\frac{\sigma_Y^2}{\sigma_Y^2}=1
\end{equation}
所以$X^\prime$与$Y^\prime$的相关系数的估计为：
\begin{equation}
r_{X^\prime Y^\prime} = \frac{\sum_{i=1}^n(X_i^\prime-\bar{X^\prime})(Y_i^\prime-\bar{Y^\prime})}{\sqrt{var(X_i^\prime)var(Y_i^\prime)}}=\sum_{i=1}^nX_i^\prime Y_i^\prime = (X^\prime)^TY^\prime
\end{equation}

残差分解：


 * 总平方和:$SST=\sum_{i=1}^n(Y_i-\bar{Y})^2$，度量因变量总的变异程度	
 * 回归平方和:$SSR=\sum_{i=1}^n(\hat{Y}_i-\bar{Y})^2$，表示的是回归模型所能够解释的那部分的变异程度
 * 残差平方和:$SSE=\sum_{i=1}^n(Y-\hat{Y}_i)^2=\sum_{i=1}^n e_i^2$，表示回归模型不能解释的那部分的变异程度
 * $SST=SSR+SSE$


可见，给定数据时总平方和是固定的，且总平方和等于回归平方和与残差平方和之和。

可见，最小二乘法就是要使残差平方和$\sum_{i=1}^n e_i^2$最小, 即使$e$的长度尽可能小。在几何上来看，当且仅当$e \perp X$时，$e$的长度最小。换句话说，$\hat{Y}$的长度是$Y$在$X$轴上的投影长度.

判定系数$R^2$。令$Y^\prime = Y-\bar{Y}$，$\hat Y^\prime =\hat Y-\bar{Y}$判定系数的几何定义为：

\begin{equation}\label{key}
R^2 =\frac{SSR}{SST}=  \frac{\|\hat Y^\prime \|^2}{\|Y^\prime \|^2} = cos^2 \theta
\end{equation}

$R^2$反映了回归模型可解释部分解释$Y$的程度。根据$\frac{\partial Q}{\beta_1}|_{\beta_1=\hat\beta_1}  = -2\sum_{i=1}^n(Y_i-\hat\beta_0-X_i\hat\beta_1)X_i=0$式的定义，$R^2$的几何含义如下图所示。

判定系数$R^2$也可以表达为：

\begin{equation}\label{key}
R^2 =\frac{SSR}{SST} = \frac{SST-SSE}{SST}  =  1-\frac{SSE}{SST} 
\end{equation}
可见，$R^2$越大，模型拟合效果越好，$R^2$越小，模型拟合效果越差.

### 9.2.2 一元线性回归模型的假设检验

**判定系数$R^2$：**<p>
$R^2$称为判定系数, 它反映了回归模型的拟合效果.
判定系数$R^2$的定义为：
$$R^2=\frac{\sum_{i=1}^n (\hat{Y_j}-\bar{Y})^2}{\sum_{i=1}^n(Y_i-\bar{Y})^2}$$
它表示被自变量解释的部分的变异性占总体变异性的比例.

注：多元回归的几何解释的图形与一元回归的几何解释图形完全相同, 不再表示一个坐标, 而是多个变量表示的一个平面.

**调整之后的判定系数：**<br>
Adjust$R^2$：即校正的判定系数，指“考虑了自由度的判定系数”，记为$R_{adjust}^2$。

\begin{equation}\label{key}
R_{adjust}^2 =1- (1-R^2) \frac{n-1}{n-k} 
\end{equation}

从$R_{adjust}^2$的定义可见，$R_{adjust}^2$剔除了自由度的影响。$R_{adjust}^2$越大，表明模型拟合效果越好，反之越差。<p>

**F统计量：**<br>
F统计量:
$$F=\frac{\sum_{i=1}^n (\hat{Y_i}-\bar{Y})^2/(k)}{\sum_{i=1}^n(Y_i-\hat{Y})^2/(n-k-1)}\sim F(k,n-k-1)$$
它是被自变量解释的部分的平均变异性与不能被自变量解释的部分的平均变异性之比。F越大，模型可解释的比例比不可解释部分大，更有可能落入拒绝域. 通过判断F是否落入拒绝域来判断模型是否显著，与方差分析一致.



F统计量与$R^2$的关系说明
$$F =\frac{SSR/(k-1)}{SSE/(n-k)} = \frac{SSR}{SSE}\cdot\frac{n-k}{k-1} =\frac{SSR}{SST-SSR}\cdot\frac{n-k}{k-1} = \frac{R^2}{1-R^2}\cdot\frac{n-k}{k-1}$$
可见$R^2$越大，F越大，模型越显著，模型拟合效果越好。$R^2$反应了模型拟合效果的好坏，而F则用来检验模型是否在统计学的角度有意义，即通过原假设成立的条件下，F的发生概率。<br>
因此，可以根据 F 的值来实现模型的假设检验，其**过程**如下：<br>
   * 第1步: 提出假设<br>
   $H_0$:$\beta_1=0$ 或 回归模型不存在<br>
   $H_1$:$\beta_1\neq0$ 或 回归模型存在<br>
   
   * 第2步: 计算检验统计量<br>
   首先，计算F统计量<br>
   
   * 第3步：进行决策<br>
   给定显著水平$\alpha$，求出统计量F的p-value值，如果p-value$<\alpha$，则拒绝$H_0$，表示线性回归模型显著。否则，接受$H_0$，表示回归模型不显著。

### 9.2.3 一元线性回归模型参数的假设检验

通常，采用t检验来检验回归系数$\beta_0$和$\beta_1$的显著性，具体原理将在多元线性回归中讲解，本小节只介绍其过程。

回归系数检验过程:
* 第1步: 提出假设<br>
	$H_0$:$\beta_j=0$ 或 回归系数不显著<br>
	$H_1$:$\beta_j\neq0$ 或 回归系数显著<br>
	
* 第2步: 计算检验统计量<br>
   首先，计算$t$统计量
	
	$$t=\frac{\hat\beta_j-\beta_j}{\hat\sigma_{\hat\beta_j}}\sim t(n-k)$$
   其中${\hat\sigma_{\hat\beta_j}}$是$\hat\beta_j$的标准误${\hat\sigma_{\hat\beta_j}}$=${\hat\sigma_{u}\sqrt{c_{jj}}}$，$c_{jj}$=${(X^TX)_{jj}}^{−1}$<br>
注意：对于一元线性回归，模型只有1个变量，但有两个参数（$\beta_0$和$\beta_1$），因此k=2
	
* 第3步：进行决策<br>
	给定显著水平$\alpha$，若统计量满足条件|$t$| > |$t_{\alpha/2}(n-k)$|, 就拒绝原假设，表示回归系数显著；否则不能拒绝原假设。




**预测：**<br>
**（1）因变量新值的点估计：**<br>
假设回归模型为: $\widehat{Y}=\widehat{\beta}_0+\widehat{\beta}_1 X$, 那么, 当 $X=X_0$ 时, $Y$ 的预测值 $\widehat{Y}_0$ 为:
$$\widehat{Y}_0=\widehat{\beta}_0+\widehat{\beta}_1 X_0 $$<br>
**（2）因变量新值的区间估计:**<br>
当 $X=X_0$ 时, $\widehat{Y}_0=\widehat{\beta}_0+\widehat{\beta}_1 X_0$, 那么可以计算 $\widehat{Y}_0$ 的期望和方差, 分别为:
$$
\begin{gathered}
E\left(\widehat{Y}_0\right)=E\left(\widehat{\beta}_0+\widehat{\beta}_1 X_0\right)=\widehat{\beta}_0+\widehat{\beta}_1 X_0 \\
\operatorname{var}\left(\widehat{Y}_0\right)=\operatorname{var}\left(\bar{Y}-\widehat{\beta}_1 \bar{X}-\widehat{\beta}_1 X_0\right)=\left[\frac{1}{n}+\frac{\left(X_0-\bar{X}\right)^2}{\sum_{i=1}^n\left(X_i-\bar{X}\right)^2}\right] \sigma^2=h_{00} \widehat{\sigma}^2
\end{gathered}
$$
因此, $\widehat{Y}_0 \sim N\left(\widehat{\beta}_0+\widehat{\beta}_1 X_0, h_{00} \sigma^2\right)$, 给定置信水平 $\alpha, \widehat{Y}_0$ 的置信区间 为:
$$
\widehat{Y}_0 \pm\left|t_{\frac{\alpha}{2}}(n-2)\right| \sqrt{1+h_{00}} \widehat{\sigma}
$$

## 9.3 多元线性回归模型：

设变量$Y$与变量$X_1$,$X_2$,$X_3$,…,$X_p$之间的线性关系
$$Y=\beta_0+\beta_1\beta_1+X_2\beta_2+…+X_p\beta_p+\epsilon$$


其中$\epsilon^{iid}\sim{N(0,{\sigma}^2)}$，$\beta_0，\beta_1，……，\beta_p$和$\sigma^2$是未知参数，当$p\geq 2$时，称上述模型为多元线性回归模型。

### 9.3.1 多元线性回归模型最小二乘估计：
$$
\widehat{\beta}=\left(X^T X\right)^{-1} X^T Y
$$
$\widehat{\beta}=(\widehat{\beta_0},\widehat{\beta_1},……,\widehat{\beta_p})是回归系数，\widehat{\beta_j}(j=1,2,……，p)$是偏回归系数。<br>
回归系数$\hat{\beta_j}$表示在其他解释变量保持不变的情况下，$X_j$每变化1个单位时，Y的均值E($Y$)的变化。<br>
或者说，$\hat{\beta_j}$给出了$X_j$单位变化对$Y$均值的“直接”或 “净”(不含其他变量) 影响。<br>

### 9.3.2 多元线性回归模型回归拟合效果分析：
* （1）回归模型的拟合效果
* （2）回归系数的显著性检验

**(1)回归方程的拟合效果:**<br>

* 判定系数:
\begin{equation}\label{key}
R^2 = \frac{SSR}{SST}
\end{equation}

其中：$SST=SSE+SSR$，$SST$是总体离差平方和且$SST=\sum_{i=1}^n(y_i-\bar{y})^2$, SSR是回归平方和且$SSR=\sum_{i=1}^n(\hat{y}_i-\bar{y})^2$


$R^2$表示回归模型可解释部分占总体的百分比，$R^2\leq1$。<br>
$R^2$越大，说明模型拟合效果越好，模型的可解释部分被拟合的程度越高。

$R_{adjust}^2$是$R^2$调整自由度后的判定系数：

\begin{equation}\label{key}
R_{adjust}^2 =1- (1-R^2) \frac{n-1}{n-k-1} 
\end{equation}


* F检验:

原假设：$H_0:\beta_1=\beta_2=...=\beta_p=0(模型不存在)$<br>
备择假设：$H_1:\beta_1,\beta_2,...,\beta_p不全为0(模型存在)$<br>
    
回归方程显著存在时，拒绝原假设；否则接受原假设，说明不存在多元线性关系。


构造F统计量：<br>
	
$$F=\frac{SSR/k}{SSE/(n-k-1)} \sim F(k,n-k-1)$$
	
其中：$SSR=\sum_{i=1}^n(\hat{y}_i-\bar{y})^2$，$SSE=\sum_{i=1}^n(y_i-\hat{y}_i)^2$<br>
$SSR$是回归平方和，$SSE$是残差平方和。对于给定的显著水平$\alpha$，检验的拒绝域为：$|F|\geq{F_{\alpha/2}(k,n−k−1)}$


**（2）回归系数的显著性检验：**<br>
原假设：$H_0:\hat{\beta}_j=0$<br>
备择假设；$H_1:\hat{\beta}_j \neq 0$<br>
如果回归系数的显著性检验的原假设成立，此时$\beta_j$=0，说明$X_j$与$Y$无关，$X_j$的变化不会引起$Y$的变化<br>
构造T统计量:<br>

$$T_j=\frac{\hat{\beta}_j}{\hat{\sigma}\sqrt{c_{jj}}} \sim t(n-k);j=1,2,……，p;k=p+1$$
$c_{jj}$是$(X^T X)^{-1}$对角线上第$j$个元素。对于给定的显著水平$\alpha$，检验的拒绝域为：$|T_j|\geq t_{\alpha/2}(n-k)$

## 9.4 非线性回归

在许多实际问题中，变量之间的关系并不都是线性的。通常我们会碰到某些现象的被解释变量与解释变量之间呈现某种曲线关系。对于曲线形式的回归问题，显然不能照搬我们前面线性回归的建模方法。本章我们首先讨论可化为线性回归的曲线回归问题，然后讨论一种分段回归方法，再讨论一般非线性回归模型的参数估计方法和建模过程。

### 9.4.1可化为线性回归的曲线回归
实际问题中, 有许多回归模型的被解释变量 $y$ 与解释变量 $x$ 之间的关系者都不是线性的,其中一些回归模型通过对自变量或因变量的函数变换可以转化为线性关系, 利用线性回归求解末知参数, 并作回归诊断。如有下列模型
$$
\begin{aligned}
& y=\beta_0+\beta_1 \mathrm{e}^x+\varepsilon \\
& y=\beta_0+\beta_1 x+\beta_2 x^2+\cdots+\beta_p x^p+\varepsilon \\
& y=a e^{b x} \mathrm{e}^{\varepsilon} \\
& y=a \mathrm{e}^{b x}+\varepsilon
\end{aligned}
$$
对于上式, 只需令 $x^{\prime}=\mathrm{e}^x$ 即可化为 $y$ 对 $x^{\prime}$ 是线性的形式
$$
y=\beta_0+\beta_1 x^{\prime}+\varepsilon
$$
需要指出的是, 新引进的自变量只能依赖于原始变量, 而不能与末知参数有关。 对于上式, 我们可以令 $x_1=x, x_2=x^2, \cdots, x_p=x^p$, 于是得到 $y$ 关于 $x_1, x_2, \cdots, x_p$ 的线性表达式
$$
y=\beta_0+\beta_1 x_1+\beta_2 x_2+\cdots+\beta_p x_p+\varepsilon
$$
上式本来只有一个自变量 $x$, 是一元 $p$ 次多项式回归, 在线性化后, 变 为 $p$ 元线性回归。
对于上式, 对等式两边同时取自然对数, 得
$$
\ln y=\ln a+b x+\varepsilon
$$
令 $y^{\prime}=\ln y, \beta_0=\ln a, \beta_1=b$, 于是得到 $y^{\prime}$ 关于 $x$ 的一元线性回归模型
$$
y^{\prime}=\beta_0+\beta_1 x+\varepsilon
$$
对于上式, 当 $b$ 末知时, 我们不能通过对等式两边同时取自然对数的方法将回归模型线性化, 只能用非线性最小二乘方法求解。<br>
回归模型$y=a e^{b x} \mathrm{e}^{\varepsilon}$式可以线性化, 而回归模型$y=a \mathrm{e}^{b x}+\varepsilon$式不可以线性化, 两个回归模型有相同的回归函数 $a \mathrm{e}^{b x}$, 只是误差项 $\varepsilon$ 的形式不同。$y=a e^{b x} \mathrm{e}^{\varepsilon}$式的误差项称为乘性误差项, $y=a \mathrm{e}^{b x}+\varepsilon$ 式的误差项称为加性误差项。因而一个非线性回归模型是否可以线性化, 不仅与回归函数的形式有关, 而且与误差项的形式有关, 误差项的形式还可以有其他多种形式。<br>
$y=a e^{b x} \mathrm{e}^{\varepsilon}$式与 $y=a \mathrm{e}^{b x}+\varepsilon$ 式的回归参数的估计值是有差异的。对误差项的形式, 首先应该由数据的经济意义来确定, 然后由回归拟合效果作检验。过去, 由于没有非线性回归软件, 人们总是希望非线性回归模型可以线性化, 因而误差项的形式还可以有其他多种形式。

### 9.4.2多项式回归
多项式回归模型是一种重要的曲线回归模型, 这种模型通常容易转化成一般的多元线性回归来作处理, 因而它的应用也十分广泛。<br>
一、几种常见的多项式回归模型<br>
回归模型<br>
$$
y_i=\beta_0+\beta_1 x_i+\beta_2 x_i^2+\varepsilon_i
$$
称为一元二阶 (或一元二次) 多项式模型。其中 $i=1,2, \cdots, n$,在以下的回归模型 中不再一一注明。<br>
为了反映回归系数所对应的自变量次数的类型, 我们通常将多项式回归模型 中的系数表示成下面模型中的情形
$$
y_i=\beta_0+\beta_1 x_i+\beta_{11} x_i^2+\varepsilon_i
$$
上式的回归函数 $y_i=\beta_0+\beta_1 x_i+\beta_{11} x_i^2$ 是一条抛物线方程, 通常称为 二项式回归函数。回归系数 $\beta_1$ 为线性效应系数, $\beta_{11}$ 为二次效应系数。<br>
相应地, 回归模型
$$
y_i=\beta_0+\beta_1 x_i+\beta_{11} x_i^2+\beta_{111} x_i^3+\varepsilon_i
$$
称为一元三次多项式模型。
当自变量的幂次超过 3 时, 回归系数的解释变得困难起来, 回归函数也变得 很不稳定, 对回归模型的应用会受到影响。因而, 常次超过 3 的多项式回归模型不常使用。以上两个多项式回归模型都是只含有一个自变量 $x$, 在实际应用中, 我们常遇到含有两个或两个以上自变量的情况。称回归模型
$$
y_i=\beta_0+\beta_1 x_{11}+\beta_2 x_{i 2}+\beta_{11} x_{11}^2+\beta_{22} x_{i 2}^2+\beta_{12} x_{i 1} x_{i 2}+\varepsilon_i
$$
为二元二阶多项式回归模型。它的回归系数中分别含有两个自变荲的线性项系数 $\beta_1$ 和 $\beta_2$, 二次项系数 $\beta_{11}$ 和 $\beta_{22}$, 并含有交叉乘积项系数 $\beta_{12}$ 。交叉乘积项表示 $x_1$与 $x_2$ 的交互作用, 系数 $\beta_{12}$ 通常称为交互影响系数。


### 9.4.3 非线性最小二乘
非线性回归模型一般可记为
$$y_i=f\left(x_i, \boldsymbol{\theta}\right)+\varepsilon_i, \quad i=1,2, \cdots, n$$
其中，$y_i$为因变量；非随机向量$\boldsymbol{x}_i=\left(x_{i 1}, x_{i 2}, \cdots, x_{i k}\right)^{\prime}$是自变量；$\boldsymbol{\theta}=\left(\theta_0\right.$, $\left.\theta_1, \cdots, \theta_p\right)^{\prime}$ 为未知参数向量；$\epsilon_i$为随机误差项并且满足独立同分布假定，即$$
\left\{\begin{array}{l}
E\left(\varepsilon_i\right)=0, \quad i=1,2, \cdots, n \\
\operatorname{cov}\left(\varepsilon_i, \varepsilon_j\right)=\left\{\begin{array}{ll}
\sigma^2, & i=j \\
0, & i \neq j
\end{array} \quad(i, j=1,2, \cdots, n)\right.
\end{array} \quad\right.
$$
如果 $f\left(\boldsymbol{x}_i, \boldsymbol{\theta}\right)=\theta_0+x_1 \theta_1+x_2 \theta_2+\cdots+x_p \theta_p$, 那么 $y_i=f\left(x_i, \boldsymbol{\theta}\right)+\varepsilon_i $式就是前面讨论的线性模型，而且必然有 $k=p$; 对于一般情况的非线性模型, 参数的数目与自变量的数目并没有一定的对应关系, 不要求 $k=p$ 。

对非线性回归模型$y_i=f\left(x_i, \boldsymbol{\theta}\right)+\varepsilon_i $式, 仍使用最小二乘法估计参数 $\boldsymbol{\theta}$, 即求使得
$$
Q(\boldsymbol{\theta})=\sum_{i=1}^n\left(y_i-f\left(x_i, \boldsymbol{\theta}\right)\right)^2
$$
达到最小的 $\hat{\theta}$, 称为 $\theta$ 的非线性最小二乘估计。在假定 $f$ 函数对参数 $\boldsymbol{\theta}$ 连续可微 时, 可以利用微分法, 建立正规方程组, 求解使 $Q(\theta)$ 达最小的 $\hat{\theta}$ 。将 $Q$ 函数对 㣍数 $\boldsymbol{\theta_j}$,求偏导, 并令其为 0 , 得 $p+1$ 个方程
$$
\begin{aligned}
& \left.\frac{\partial Q}{\partial \theta_j}\right|_{\theta_{j}=\hat{\theta}_{j}}=-\left.2 \sum_{i=1}^n\left(y_i-f\left(x_i, \hat{\theta}\right)\right) \frac{\partial f}{\partial \theta_j}\right|_{\theta_j=\hat{\theta_j}}=0 
& j=0,1,2, \cdots, p
\end{aligned}
$$

非线性最小二乘估计 $\boldsymbol{\hat{\theta}}$ 就是上式的解, 上式称为非线性最小二乘 估计的正规方程组, 它是末知参数的非线性方程组。一般用 Newton 迭代法求解 此正规方程组。<br>
也可以直接极小化残差平方和 $Q(\boldsymbol{\theta})$, 求出末知参数$\theta$的非线性最小二乘估计$\hat{\boldsymbol{\theta}}$ 。<br>
对于非线性最小二乘估计，我们仍然需要作参数的区间估计、显著性检 验、回归方程的显著性检验等回归诊断,这需要知道有关统计量的分布。在 非线性最小二乘中, 一些精确分布是很难得到的, 在大样本时, 可以得到近似的分布。计算机软件在求出参数 $\theta$ 的非线性最小二乘估计值 $\hat{\theta}$ 的同时，还给出近似的参数的区间估计、显著性检验、回归方程的显著性检验等回归诊断。

在非线性回归中, 平方和分解式$ \mathrm{SST}=\mathrm{SSR}+\mathrm{SSE}$ 不再成立。类似于线性回 归中的复决定系数, 定义非线性回归的相关指数：
$$
R^2=1-\frac{\mathrm{SSE}}{\mathrm{SST}}
$$

## 9.5 变量选择

回归自变量的选择无疑是建立回归模型的一个极为重要的问题。在建立一个实际问题的回归模型时，我们首先碰到的问题便是如何确定回归自变量，一般情况下，我们大都是根据所研究问题的目的，结合经济理论罗列出对因变量可能有影响的一些因素作为自变量。如果我们遗漏了某些重要的变量，回归方程的效果肯定不会好;如果我们担心遗漏了重要的变量，而考虑过多的自变量，在这些变量中，某些自变量对问题的研究可能并不重要，有些自变量数据的质量可能很差，有些变量可能和其他变量有很大程度的重叠，如果回归模型把这样一些变量都选进来，不仅计算量增大许多，而且得到的回归方程稳定性也很差，直接影响到回归方程的应用。 

### 9.5.1 多重共线性产生的原因

在线性回归中，如果自变量$X_1$, $X_2$, $X_3$,…,$X_p$之间存在高度相关，称回归模型中存在**多重共线性**，多重共线性会影响回归系数的估计结果，可能会导致：<br>
（1）回归系数不显著<br>
（2）回归系数正负号颠倒<br>
**多重共线性产生的背景和原因：**<br>
解释变量之间完全不相关的情形是非常少见的，尤其是研究某个经济问题时，涉及的自变量较多，我们很难找到一组自变量，它们之间互不相关，而且它们又都对因变量有显著影响。客观地说，某一经济现象，涉及多个影响因素时，这多个影响因素之间大都有一定的相关性。当它们之间的相关性较弱时，我们一般就认为符合多元线性回归模型设计矩阵的要求;当这一组变量间有较强的相关性时，就认为是一种违背多元线性回归模型基本假设的情形。<br>
当所研究的经济问题涉及时间序列资料时，由于经济变量随时间往往存在共同的变化趋势，使得它们之间容易出现共线性。例如，我国近年来的经济增长态势很好，经济增长对各种经济现象都产生影响，使得多种经济指标相互密切关联。比如要研究我国居民消费状况，影响居民消费的因素很多，一般有职工平均工资、农民平均收入、银行利率、全国零售物价指数、国债利率、货币发行量、储蓄额、前期消费额等，这些因素显然既对居民消费产生重要影响，它们之间又有着很强的相关性。<br>
对于许多利用截面数据建立回归方程的问题常常也存在自变量高度相关的情形。例如，以企业的截面数据为样本估计生产函数，由于投人要素资本$K$、劳动力投入$L$、科技投入$S$、能源供应$E$等都与企业的生产规模有关，所以它们之间存在较强的相关性。<br>
再如，有人在建立某地区粮食产量回归模型时，以粮食产量为因变量$y$，以化肥用量$x_1$，水浇地面积$x_2$，农业投入资金$x_3$等作为自变量。从表面上我们看到$x_1$，$x_2$，$x_3$都是影响粮食产量$y$的重要因素，可是建立的$y$关于$x_1$，$x_2$，$x_3$的回归方程效果很差，原因如何呢?后来发现尽管所选自变量$x_1$，$x_2$，$x_3$都是影响因变量y的重要因素，但是农业资金投入$x_3$与化肥用量$x_1$、水浇地面积$x_2$有很强的相关性，农业投入资金主要用于购买化肥和开发水利，也就是说资金投入的效应已被化肥用量和水浇地面积体现出来。进一步计算$x_3$分别与$x_1$,$x_2$的简单相关系数得$r_{13}$=0.98、$r_{23}$=0.99，呈现高度相关。剔除$x_3$后重新建立回归模型，结果无论从预测和结构分析来看都十分理想。 <br>

### 9.5.2 多重共线性判别方法
**(1)相关系数检验：通过相关系数和相关性检验来判断参数之前是否存在共线性。**<br>
相关性检验的假设为：<br>
$H_0:$ 两个变量之间无关 <br>
$H_1:$ 两个变量之间相关<p>
**(2)考察回归系数的显著性：通过考察回归系数的显著性来判断是否存在多重共线性。**<br>
当模型的F检验显著时，多个回归系数的t检验却不显著，表示模型可能存在多重共线性.<p>
**(3) 回归参数的正负号。**<br>
此外，也可以通过观察回归系数的正负号与预期的正负号是否相反，若相反，则表示模型可能存在多重共线性。<p>
**(4)用容忍度(tolerance)和方差膨胀因子(variance inflation factor, VIF)来识别共线性。**<br>
某个自变量的容忍度等于1减去该变量作为因变量而其余变量作为自变量时所得到的线性回归模型的决定系数，即:
\begin{equation}\label{key}
1-R^2_j
\end{equation}
其中，$R^2_j$是下面线性回归模型的决定系数：
$$X_j=\gamma_0+X_1\gamma_1+X_2\gamma_2+...+X_{j-1}\gamma_{j-1}+X_{j+1}\gamma_{j+1}+...+X_{k}\gamma_{k}$$
方差膨胀因子等于容忍系数的倒数，即：
$$VIF = \frac{1}{1-R^2_i}$$
通常情况下，VIF大于10时存在严重多重共线性。<p>

### 9.5.3 多重共线性消除方法：逐步回归
并不是所有自变量都对因变量$y$有显著的影响，这就存在着如何挑选出对因变量有显著影响的自变量问题。自变量的所有可能子集构成$2^n-1$个回归方程，当可供选择的自变量不太多时，用前边的方法可以求出一切可能的回归方程，然后用几个选元准则去挑出最优的方程，但是当自变量的个数较多时，要求出所有可能的回归方程是非常困难的。为此，人们提出了一些较为简便、实用、快速的选择最优方程的方法。人们所给出的方法各有优缺点，至今还没有绝对最优的方法，目前常用的方法有“前进法”“后退法”、“逐步回归法”，而逐步回归法最受推崇。<br>
**（1）前进法：**前进法的思想是变量由少到多，每次增加一个，直至没有可引入的变量为止。<br>
第一步，拟合因变量$Y$对k个自变量$(X_1,X_2,...,X_k)$的一元线性回归模型，共k个，然后找出回归模型拟合效果判别指标最佳的模型。将最佳模型的自变量$X_j$放入线性回归模型中。<br>
第二步，在第一步已初始化的模型基础上，引入其他自变量。拟合因变量$Y$和被选入的自变量之间的多元线性回归，计算回归模型拟合效果判别指标，选择回归模型拟合效果判别指标最优的线性回归模型。若加入新变量后，模型的拟合效果较初始模型有所改善，则把新变量放入到线性回归模型中。<br>
第三步，依次引入新指标，直到线性回归模型的拟合效果没有任何改善为止。<p>
**（2）后退法：**后退法与前进法相反，首先用全部k个变量建立一个回归方程，然后在这k个变量中选择一个最不重要的变量，将它从方程中剔除。<br>
第一步，拟合因变量$Y$对k个自变量$(X_1,X_2,...,X_k)$的多元线性回归模型。<br>
第二步，在初始模型的基础上，剔除其中一个变量$X_j$，并拟合因变量与剩余自变量之间的多元回归模型，计算回归模型拟合效果判别指标，若剔除变量后回归模型的拟合效果有改善且改善效果最好，那么剔除对应的变量。<br>
第三步，依次剔除其他变量，直到模型没有任何改善为止。<p>
前进法和后退法显然都有明显的不足。前进法可能存在这样的问题，即不能反映引进新的自变量后的变化情况。因为某个自变量开始可能是显著的，但当引入其他自变量后它变得并不显著了，但是也没有机会将其剔除，即一旦引入，就是“终身制”的；这种只考虑引入，而没有考虑剔除的做法显然是不全面的。而且，我们会发现可能最先引入的某个自变量，当其他自变量相继引入后，它会变得对因变量$y$很不显著。<br>
后退法的明显不足是，一开始把全部自变量引入回归方程，这样计算量很大。如果有些自变量不太重要，一开始就不引入，就可减少一些计算量;再就是一旦某个
自变量被剔除，“一棍子就把它打死了”，它再也没有机会重新进人回归方程。<br>
如果说我们的问题涉及的自变量$x_1，x_2，…，x_m$是完全独立的(或不相关)，那么在取$\alpha_进=\alpha_出$（$\alpha_进$:引入自变量的显著性水平;$\alpha_出$:剔除自变量的显著性水平）时，前进法与后退法所建的回归方程是相同的。然而在实际中很难碰到自变量间真正无关的情况，尤其是经济问题中，我们所研究的绝大部分问题，自变量间都有一定的相关性。这就会使得随着回归方程中变量的增加和减少，某些自变量对回归方程的影响也会发生变化。这是因为自变量间的不同组合，由于它们相关的原因，对因变量$y$的影响可能大不一样。如果几个自变量的联合效应对$y$有重要作用，但是单个自变量对$y$的作用都不显著，那么前进法就不能引引入这几个自变量，而后退法却可以保留这几个自变量，这是后退法的一个优点。<br>
从前进法和后退法的思想及方法，以及我们看到的它们的不足，人们比较自然地想构造一种方法，即吸收前进法和后退法的优点，克服它们的不足，把两者结合起来，这就产生了逐步回归法。<br>
**（3）逐步回归法：**逐步回归的基本思想是有进有出。<br>
具体做法是将变量一个一个引入，当每引入一个自变量后，对已选入的变量要进行逐个检验，当原引入的变量由于后面变量的引入而变得不再显著时，要将其剔除。引入一个变量或从回归方程中剔除-个变量，为逐步回归的一步，每一步都要进行$F$检验，以确保每次引入新的变量之前回归方程中只包含显著的变量。这个过程反复进行，直到既无显著的自变量选入回归方程，也无不显著自变量从回归方程中剔除为止。这样就避免了前进法和后退法各自的缺陷，保证了最后所得的回归子集是最优回归子集。<br>
在逐步回归法中需要注意的一个问题是引人自变量和剔除自变量的显著性水平$\alpha$值是不相同的，要求引入自变量的显著性水平$\alpha_进$小于剔除自变量的显著性水平$\alpha_出$，否则可能产生“死循环”。也就是当$\alpha_进$>$\alpha_出$时，如果某个自变量的显著性$P$值在$\alpha_进$与$\alpha_出$之间，那么这个自变量将被引入、剔除，再引入、再剔除，循环往复，以至无穷。

**逐步回归法实战：**<br>

### 9.5.4 模型比较

在回归模型中，希望所建立的模型能够更好地拟合数据，同时又不希望模型包含一些不必要的自变量。虽然逐步回归能够消除多重共线性带来的影响，但是逐步回归不能够说明两个模型之间是否有足够的统计差异。这就需要我们对不同的线性回归模型进行比较，以判断选择那些变量构建的回归模型比其他回归模型的拟合效果更好。<br>
在回归模型中，如果一个模型包含另外一个模型的所有项，并且至少有一个额外项，称这两个模型是嵌套模型(nested model)。例如只包含全部自变量$X_1,X_2,...,X_k$的回归模型，与包含交互项及二次项的回归模型$X_iX_j$及$X^2_i$，分别定义为：<br>
模型1：$E(Y)=\beta_0+X_1\beta_1+X_2\beta_2$ <br>
模型2：$E(Y)=\beta_0+X_1\beta_1+X_2\beta_2+X_1X_2\beta_3+X_1^2\beta_4$<br>
模型2包含模型1的所有项，因此模型2称为完全模型(full model)；模型1相对模型2更简单，称为简化模型(reduced model)。完全模型是否比简化模型提供更多的信息呢？如果要比较两个模型的差异性，可以做如下假设检验：<br> 
$H_0:\beta_3=\beta_4=0$ 即：模型1与模型2相同 或 模型中的二次项不会对$Y$的预测提供额外信息<br> 
$H_1:\beta_3,\beta_4$其中之一不为0，即：模型1与模型2不相同 或 模型中的二次项会对$Y$的预测提供额外信息<br> 
如何构造统计量来实现假设检验呢？检验过程如下：<br>
首先，用最小二乘法拟合简化模型，并计算相应的残差平方和($SSE_R$)；同理，拟合完全模型，并计算它们的残差平方和($SSE_F$)<br>
然后，计算出二者之差($SSE_R-SSE_F$)，并与$SSE_F$进行比较；如果二次项对模型有贡献，差值较大，说明完全模型提供了更多的信息，否则提供的信息不够多。根据这一原理，构造F统计量，如下：<br>
\begin{equation}\label{key}
F=\frac{(SSE_R-SSE_F)/(k-g)}{SSE_F/(n-k-1)}\sim F(k-g,n-k-1)
\end{equation}
最后，计算F统计量的p-value，并与给定显著水平$\alpha$进行比较，推断是否拒绝原假设；若拒绝原假设，则表示全模型和部分模型无差异，则应使用部分模型。
