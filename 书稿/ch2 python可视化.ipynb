（4）List（列表型）

列表是Python中使用最频繁的数据容器之一。列表与C语言中的数组相似，但列表的功能比数组强大. 列表是一种复合的数据类型，其数据项不需要具有相同的类型，因此列表可以用于存储数字、字符串、列表等单一或混合元素的集合。

列表包括列表的名称和列表的值，列表中的值用中括号来定义，元素之间通过逗号隔开。列表中的元素可以是字符串、数字，或者是列表等元素。

例如，定义列表t，它包括五个字符，分别是a,b,c,d,e，用逗号隔开。



列表的下标与字符串类似，列表的索引从左开始是0，最大下标为（列表长度-1）；从右开始是-1，最小下标为（-列表长度）。

**列表的访问**

列表的访问规则与字符串的访问规则相同，可以通过列表名下标来进行访问。例如，取t中的第二到第三个元素，可以用1冒号3来读取；取第一个到第四个元素，用空冒号4来读取；取第四个到最后一个元素，用3冒号空来取；取所有元素，用空冒号空来读取。大家注意，如果冒号之前为空，表示从头开始取；如果冒号之后为空，表示取到最后一个元素。注意：如果要计算列表的长度，可以通过len函数来获取。



````python
#列表
t = ["a","b","c","d","e"]
#打印列表的长度
print(len(t))
````

除了通过下标来读取，还可以通过for来逐个访问列表中的元素。For循环的语法如下，包括for语句，i是循环索引值，list1是列表，从头开始循环，每循环一次，都会判断i是否在列表中，因此i就从头开始变量list1中的元素。

````python
#方法一：通过for循环访问
list1 = [2,3,4]
for i in list1:
   print(i)
````

（5）Tuple（元组）

元组（tuple）与列表（list）相似，不同之处在于元组的元素不能修改，它相当于是只读的列表。元组是通过小括号来定义的，也包括元组名和存放在元组中的数据。与列表相似，元组也可以存储字符串、数字、列表等元素，并且元素可以不相同。

**元组的定义**：元组名 = (元素1，元素2，元素3，…)

**元组的访问**

元组的访问和列表的访问相同，都是通过下标来访问的，例如访问元组tup1的第一、第二到第四个元素的操作如下。值得注意的是：因为元组是只读的，所以不能修改元组及元组中的元素，这种操作是会报错的。

````python
#元组的定义与访问
tup1 = ( 'runoob', 786 , 2.23, 'john', 70.2 )
tup2 = (123, 'john')
 
print(tup1)               # 输出完整元组
print(tup1[0])            # 输出元组的第一个元素
print(tup1[1:3])          # 输出第二个至第四个（不包含）的元素 
print(tup1[2:])           # 输出从第三个开始至列表末尾的所有元素
````

**元组的运算**

同理，我们也可以用加和乘来组合元组、重复元组，例如把元组1重复两次，把元组1和元组2组合在一起。

````python
#元组的定义与访问
tup1 = ( 'runoob', 786 , 2.23, 'john', 70.2 )
tup2 = (123, 'john')
 
print(tup1 * 2)           # 输出元组两次
print(tup1 + tup2)        # 打印组合的元组
````

（6）Dictionary（字典）

字典（dictionary）是除列表以外Python之中最灵活的内置数据结构类型。字典用“{ }”来定义，字典由键（key）和值（value）成对组成。字典的键是唯一的，值可以不唯一，且值可以是任何python对象，如字符串、数字、列表、元组等。

字典与列表的区别：列表是有序的对象集合，字典是无序的对象集合；字典中的元素是通过键来访问的，而不是通过下标来访问。

**字典的定义**：字典名 = {key1:value1,key2:value2,key3:value3…}；

**字典的访问**

我们可以通过中括号操作来访问字典中的元素，中括号中的元素为键。例如访问字典中键为'one'的元素，访问键为2的元素，结果如下。

````python
#字典的定义与访问
dict = {}       
dict['one'] = "This is one"
dict[2] = "This is two"

print(dict['one'])           # 输出键为'one' 的值
print(dict[2])               # 输出键为 2 的值
````

**字典的修改与删除**

字典的修改也是通过赋值来实现的，具体操作就是以键为索引找到要修改的字典的元素，对其赋值，则可以实现修改。例如修改字典中，‘age’键对应的值，将其改为18。

````python
dict = {'name':'王海','age':17,'class':'计算机一班'}
dict['age'] = 18             # 修改元素
print(dict)
````

我们可以通过del操作来删除字典中的元素，例如删除‘age’键对应的值；如果想要删除字典中的所有值，则可以通过clear操作来实现。

````python
dict = {'name':'王海','age':17,'class':'计算机一班'}
del dict['age']              # 删除元素
print(dict)
dict.clear()                 # 删除所有元素
print(dict)
````

**字典的运算**

我们可以通过in方法来判断键是否出现在字典中，如果出现返回True，否则返回False。

````python
dict = {'name':'王海','age':17,'class':'计算机一班'}
print('name' in dict )
````

如果要获取字典中的所有值，则可以通过values()方法来返回。

````python
dict = {'name':'王海','age':17,'class':'计算机一班'}
print(dict.values())
````

如果要遍历字典中的元素：可以通过items()函数来返回字典中的所有键值对，再进行访问。例如，通过for循环来遍历字典中的键值对，其操作如下所示。在该操作中，items()方法把字典中每对key和values组成一个元组，并把这些元组放在列表中返回。

````python
tinydict = {'name':'runoob','code':6734, 'dept': 'sales'}

for key,values in tinydict.items():
    print(key,values)
````

（7）Set（集合）

在python中，集合是一个无序的不重复序列，集合中不会有重复的元素。我们可以使用大括号{}或set()函数来创建集合，set的输入是一个元组。例如，创建集合t，创建集合w。用于创建w的是一个元组，因为它们是不重复的序列，因此集合会自动删除重复的元素。

````python
#集合
s1 = {"Hello","world","world"}
print(s1)
w = ("Hello","world","world")
s2 = set(w)
print(s2)
````

对集合的运算主要有差集、并集和交集运算，主要是通过减号、竖线和and符号实现集合的差集、并集和交集运算，例如：集合1和集合2的差集为set1-set2；集合1和集合2的并集为set1|set2；集合1和集合2的交集为set1&set2。

````python
s1 = set('abcd')
s2 = set('cdef')

print(s1)
print(s2)
print('s1和s2的差集',s1-s2)
print('s1和s2的并集',s1|s2)
print('s1和s2的交集',s1&s2)
````

（1）条件语句

Python的if语句的功能跟其他语言的非常相似，都是用来判断给出的条件是否满足，然后根据判断的结果决定是否执行给出的操作。



它由三个部分组成，包括：关键字if本身，测试条件真假的表达式和表达式的结果为真时要执行的代码。下面是条件语句的实例，在该语句中，我们通过判断name是否等于给定的名称python，如果等于则把flag修改为True，并打印welcome boss；否则，打印名字。因为初始条件为False，条件不成立，则执行否则语句，打印luren。

````python
flag = False
name = 'luren'
if name == 'python':         # 判断变量是否为 python 
    flag = True              # 条件成立时设置标志为真
    print('welcome boss')     # 并输出欢迎信息
else:
    print(name)                # 条件不成立时输出变量名称
````

Python语言也可以定义多重条件语句，通过在if之后增加elif来增加判断条件分支，从而实现多重条件语句。相关实例如下所示。

````python
num = 5     
if num == 3:             # 判断num的值
    print('boss')        
elif num == 2:
    print('user')
elif num == 1:
    print('worker')
elif num < 0:            # 值小于零时输出
    print('error')
else:
    print('roadman')      # 条件均不成立时输出
````

**While循环**

Python编程中while语句用于循环执行程序，即在某条件下循环执行某段程序，以处理需要重复处理的相同任务。流程如图所示。



例如：循环打印，每次执行循环体，都会判断count是否小于9，如果count小于九，则会执行循环体内的程序代码，打印并且count会加一，再次循环。直到count不小于九为止。

````python
count = 0
while (count < 9):
   print('The count is:',count)
   count = count + 1
 
print("Good bye!") 
````

While语句也可以搭配else语句一起使用，当结束循环后，会执行else语句，例如下面实例在while执行结束后，会执行else分支语句，打印相关的内容。

````python
#while...else
count = 0
while count < 5:
   print(count, " is  less than 5")
   count = count + 1
else:
   print(count, " is not less than 5")
````

### 2.1.3 函数

在python程序开发过程中。当需要重复完成某些任务时，例如：求一个商品的使用人数、满意度等统计信息。我们会根据不同商品，不断重复执行该代码块，这样会造成代码的重复率高，应用程序代码冗余、琐碎。解决这个问题的方法就是使用函数。与其他语言相似，函数在python中扮演着至关重要的角色。
在python中，我们可以将完成某一特定功能的代码编写成函数，放在函数库中供大家重复使用。开发人员要善于使用函数，以提高编码的效率，减少编写代码的工作量。
在python语言中，函数包括：内建函数，例如python的内置函数print()，len()等；以及自定义函数，用户根据自己的需求定义的函数。
在python语言中，函数的定义是通过关键字DEF来定义的，紧接着是函数的名称，函数的参数，以及函数体和返回值，具体语法如下。

def functionname( parameters ):<br>
&emsp;&emsp;"函数_文档字符串"<br>
&emsp;&emsp;function_suite<br>
&emsp;&emsp;return [expression]<br>

函数代码块以“def”关键词开头，后接函数标识符名称；任何传入参数和自变量必须放在圆括号中，参数可以是零到多个；函数体以冒号起始，是函数中缩进的部分；函数的返回值用return [表达式]返回，可以返回零到多个值；注意：参数和返回值均可以不限制类型

例如，我们下面定义一个get_stat函数。该函数的名字为get_stat,  pr是其传入的参数。函数体中包括：样本量和计算平均满意度的语句；返回值为样本量和平均满意度。在该函数中，我们调用了Python的内置函数len(),以及来自numpy程序包的mean()函数，len用来计算样本量，mean用来计算序列的平均值。Return返回值包括两个，样本量n和满意度平均值。

当我们完成函数的定义之后，我们就可以重复调用该函数。例如，计算产品1的样本量和平均值，调用方法是使用函数名，在括号中输入参数，该参数就是我们定义的序列product1，调用完后，输出n1，和mean1; 同理可以重复调用该函数计算产品2的样本量和平均值。结果通过print函数打印出来。

结果显示，product1的样本量为9，product2的样本量为11，平均满意度分别约为：2.22和3.91；说明产品2的平均满意度较高。

````python
import numpy as np

def get_stat (pr):
    n = len(pr)
    mean = np.mean(pr)
    return n, mean

product1 = [1,2,1,3,5,2,2,2,4]
product2 = [4,5,3,3,4,5,4,4,2,4,4]
n1, mean1 = get_stat(product1) 
n2, mean2 = get_stat(product2) 
print("product1: 样本量 =",n1,",满意度 =",mean1)
print("product2: 样本量 =",n2,",满意度 =",mean2)
````

我们已经学习了如何通过定义和使用函数来构造程序，但是由大量函数组成的程序通常很难理解和更新。为了解决这个问题，计算机科学家发明了面向对象的程序设计(OOP, object-oriented programming)，这是一种通过对象协作来解决任务的程序设计方式。每个对象都有自己的一组数据，以及一组作用于操作数据的方法。当开发面向对象的程序时，可以创建自定义对象来描述应用程序中的重要内容。例如，在学生数据库中，我们可以使用Student(学生)对象和Course(课程)对象，并为这些对象提供方法。

````python
"hello world".upper()
````

类提供的所有方法的集合以及它们的行为描述，称为类的公共接口(public interface)。使用类的对象时，我们并不知道该对象如何存储数据，也不知道如何实现方法。我们不需要知道str对象是如何组织字符序列的，也不需要知道列表如何存储元素。我们只需要知道公共接口，通过公共接口可以应用哪些方法，以及这些方法的作用。在隐藏实现细节的同时提供公共接口的过程称为封装(encapsulation)。

我们可以使用封装来设计自定义类。也就是说，我们将提供一组公共方法并隐藏实现的详细信息。团队中的其他程序员就可以直接使用我们定义的类而不必知道类的具体实现，就像我们可以使用str类和list类一样。如果我们参与一个长期开发的项目，实现细节通常会发生更改，一般是为了提升对象的效率或者增加对象的功能。封装对于这些更改至关重要。当具体的实现细节被隐藏时，这些改进不会影响使用对象的程序员。

（2）类的定义 ——【2】

人们在认识客观世界时经常采用抽象方法来对客观世界的众多事物进行归纳、分类。我们用类来抽象、描述待求解问题所涉及的事物，具体包括两个方面的抽象:数据抽象和行为抽象。数据抽象描述某类对象共有的属性或状态；行为抽象描述某类对象共有的行为或功能特征。

如上文所说，在Python中，使用类来定义同一种类型的对象。类(class)是广义的数据类型，能够定义复杂数据的特性，包括静态特性(即数据抽象)和动态特性(即行为抽象，也就是对数据的操作方法)。一个Python类使用变量存储数据域，称为类中的属性；定义方法来完成动作。对象是类的一个实例，一个类可以创建多个对象。创建类的一个实例的过程被称为实例化。在术语中，对象和实例经常是可以互换的。对象就是实例，实例就是对象。类和对象的关系相当于普通数据类型和它的变量之间的关系。比如可以定义一个鸟类，那么一只鹦鹉就是这个鸟类的一个对象，动物园里的一只八哥也是这个鸟类的一个对象。

类和对象的关系：
* 类是对象的抽象，而对象是类的具体实例
* 类是抽象的，而对象是具体的
* 每一个对象都是某一个类的实例
* 每一个类在某一时刻都有零或更多的实例
* 类是静态的，它们的存在、语义和关系在程序执行前就已经定义好了；对象是动态的，它们在程序执行时可以被创建和删除
* 类是生成对象的模板。

Python中使用class关键字来定义类，类名的首字母一般要大写。形如：

class Class_Name:<br>
&emsp;&emsp;属性1<br>
&emsp;&emsp;......<br>
&emsp;&emsp;属性n<br>
&emsp;&emsp;<方法定义1><br>
&emsp;&emsp;......<br>
&emsp;&emsp;<方法定义n><br>

类属性是在类中方法之外定义的，类属性属于类，可通过类名访问(尽管也可通过对象访问，但不建议这样做，因为这样做会造成类属性值不一致)。一个类中有两个特殊的方法：\_\_new\_\_()和\_\_init\_\_()。这两个方法合起来类似于Java和C++等程序设计语言中的构造方法，用于创建并初始化一个对象。当实例化(创建)一个类对象时，最先被调用的是\_\_new\_\_()方法。\_\_new\_\_()方法用于创建对象，再将该对象传递给\_\_init\_\_()方法中的self参数。而\_\_init\_\_()方法是在对象创建完成之后初始化对象状态。这两个方法都是在实例化对象时被自动调用的，不需要程序显式调用。如果用户未设计\_\_init\_\_()方法，Python将提供一个默认的\_\_init\_\_()方法。每个方法其实都是一个函数定义，与普通函数略有差别。

* 每个实例方法的第一个参数都是self，self代表将来要创建的对象实例本身。在访问类的实例属性时需要以self为前缀。
* 实例方法只能通过对象来调用，即向对象发消息请求对象执行某个方法。

下面举例定义学生类，并创建实例：

````python
#定义学生类
class Student:
    have_teacher = True
    def __init__(self, id):
        self.id = id
    def getId(self):
        print('这个学生id是：', self.id)

print(Student.have_teacher)   #输出类属性
Student.have_teacher = False  #修改类属性
print(Student.have_teacher)

stu = Student(1)              #初始化实例
stu.getId()                   #调用实例方法
stu.id = 2                    #修改实例属性
stu.getId()
````

（3）捕获与处理异常——【1】

try/except语句用来检测 try 语块中的异常，让 except 语句捕获异常信息并处理。捕获与处理异常的语法如下:


try:<br>
&emsp;&emsp;<可能出现异常的语句块><br>
except<异常类名字 name1>:<br>
&emsp;&emsp;<异常处理语句块1>&emsp;&emsp;&emsp;&emsp;#如果在 try 部分引发了 name1 异常,执行这部分语句块<br>
except<异常类名字name2>as e1:<br>
&emsp;&emsp;<异常处理语句块2>&emsp;&emsp;&emsp;&emsp;#如果在 try 部分引发了 name2 异常,执行这部分语句块<br>
except<(异常类名字name3, 异常类名字name4, ...)>as e2:<br>
&emsp;&emsp;<异常处理语句块3>&emsp;&emsp;&emsp;&emsp;#如果引发了 name3, name4, ...中任何一个异常,执行该语句块<br>
...<br>
except:<br>
&emsp;&emsp;<异常处理语句块n>&emsp;&emsp;&emsp;&emsp;#如果引发了异常,但与上述异常都不匹配,执行此语句块<br>
else:<br>
&emsp;&emsp;<else 语句块>&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;#如果没有上述所列的异常发生,执行 else 语句块<br>
finally:<br>
&emsp;&emsp;<任何情况下都要执行的语句块><br>


try中的语句块先执行。如果 try 语句块中的某一语句执行时发生异常，Python 就跳到 except 部分，从上到下判断抛出的异常对象是否与 except 后面的异常类相匹配，并执行第一个匹配该异常的 except 后面的语句块，异常处理完毕。

如果异常发生了，但是没有找到匹配的异常类别，则执行不带任何匹配类型的 except 语句后面的语句块，异常处理完毕。

如果 try 语句块的某一语句里发生了异常，却没有匹配的 except 子句，也没有不带匹配类型的 except 部分，则异常将往上被递交到上一层的 try/except 语句进行异常处理，或者直到将异常传递给程序的最上层，从而结束程序。

如果 try 语句块中的任何语句在执行时没有发生异常，Python 将执行 else 语句后的语句块。

执行完 except 后的异常处理语句或 else 后面的语句块后，程序一定会执行 finally 后面的语句块。这里的语句块主要用来进行收尾操作，无论是否出现异常都将被执行。

一个异常处理模块至少有一个 try 和一个 except 语句块，else 和 finally 语句块是可选的。

### 2.2.2 ndarray的索引与切片

现在我们已经了解了NumPy数组是什么，让我们更深入地研究这个基于索引和切片的有用的数据结构

(1)索引

和列表一样，在数组中索引尤其简单。需要注意是Python从0开始计数，使得第一个元素的索引为0。这与R不同，R从1开始计数。

````python
# start with a 1d array
rng = np.random.default_rng()
product_sales = rng.integers(low=1, high=50, size=100, endpoint=True)
````

找到第一个元素:

````python
idx = 0
product_1st = product_sales[idx]
print('Product sales of the 1st one:', product_1st)
````

我们也可以同时索引多个条目

````python
# The first and the fourth (0 -> 1st; 3 -> 4th in Python!)
idx_list = [0, 3]
product_1st_4th = product_sales[idx_list]
print('Product sales of the 1st and the 4th ones:', product_1st_4th)
````

更常见的是，我们可能会遇到二维数组（如数据库中的表，其中每行是一条记录，每列是该记录的一个属性）。例如，假设我们有以下数据：

````python
# simulate a 2d array where
# rows are user information
# columns are
### user tenure (days since they registered on the platform)
### age
### points (points made from their community engagement
rng = np.random.default_rng()
user_info = rng.integers(low=1, high=70, size=60, endpoint=True)
# Currently, does `user_info` is in the right shape?
print('user_info:\n', user_info)
````

我们如何转换形状，使这个“平面”数组成为一个二维数组?这是我们应该看数组形态的地方:

````python
print('The shape of user_info is:', user_info.shape)
````

在这个例子中，我们实际上有20个用户，每个用户有3个属性（因此我们总共有20x3＝60个元素）。我们可以按如下方式变换数组：

````python
num_users = 20
num_attr = 3
# the first argument is the number of rows; the 2nd one is the number of columns
user_info_2d = user_info.reshape(num_users, num_attr)
print('The shape of user_info_2d is:', user_info_2d.shape)
print(user_info_2d)
````

通常，我们可以直接更改`user_info_2d`的`shape`属性：

````python
user_info_2d = user_info.copy()
print('Before changing the shape:', user_info_2d.shape)
print(user_info_2d)
user_info_2d.shape = [num_users, num_attr]
print('After changing the shape:', user_info_2d.shape)
print(user_info_2d)
````

现在我们有了一个二维数组，我们如何定位一个特定的元素?比如说，我们想要检索第10个用户的年龄(第10行，第2列)

````python
idx_row = 9 # the 10th row has the index value of 9
idx_col = 1 # the 2nd column has the index value of 1
# we can index this element by inputting the indices seperated by comma
user_10th_age = user_info_2d[idx_row, idx_col]
print('The age of the 10th user is:', user_10th_age)
````

另一种方法是索引两次，我们首先提取用户的行，然后提取该用户中的特定元素（年龄）：

````python
user_10th = user_info_2d[idx_row]
print('The 10th user:', user_10th)
user_10th_age = user_10th[idx_col]
print('The 10th user\'s age:', user_10th_age)
````

````python
# This can also be done in one step
user_10th_age = user_info_2d[idx_row][idx_col]
print('The 10th user\'s age via the one-step approach:', user_10th_age)
````

有鉴于此，我们可以轻松地为任意结构中的数组元素建立索引：

尝试建立一个三维数组：这也很常见——通过添加一个时间元素，可以将之前的二维用户信息扩展到三维，这样的三维数组由三个维度组成：
- 第一层是时间（第1周、第2周等）
- 第二层是特定时间的用户（用户1、用户2等）
- 第三层是特定用户在特定时间的属性（任期、年龄、分数等）

````python
# Random number generator
rng = np.random.default_rng()
num_weeks = 4
num_users = 10
num_attr = 3
# The total number of observations is 3x10=30 in each time unit (i.e., week)
num_size = num_users * num_attr
````

````python
# Let's do this using list comprehension:
user_info_3d = np.array([rng.integers(low=1, high=70, size=num_size, endpoint=True).reshape(num_users, num_attr) for t in np.arange(num_weeks)])
# How do you interpret the shape of this array?
print('user_info_3d shape is:', user_info_3d.shape)
print(user_info_3d)
````

我们如何检索第4周 用户7的点数?

````python
# multiple indices simultaneously
user_7th_w4_points = user_info_3d[3, 6, 2]
print('User 7\'s point record in week 4 is:', user_7th_w4_points)
````

````python
# Or, we can retrieve values one at a time 
w4 = user_info_3d[3]
print('Data in week 4:\n', w4)
user_7th_w4 = w4[6]
print('Data of the 7th user in week 4:', user_7th_w4)
user_7th_w4_points = user_7th_w4[2]
print('Point of the 7th user in week 4:', user_7th_w4_points)
````

````python
# the previous method is equivalent to:
user_7th_w4_points = user_info_3d[3][6][2]
print('User 7\'s point record in week 4 is:', user_7th_w4_points)
````

最后，注意，只要元素的总数正确合适，我们可以轻松地将Numpy数组转换为任何想要的形态:

````python
# use the `size` property. Think: why are parantheses not needed here?
print('The total number of observations in user_info_3d:', user_info_3d.size)
````

(2)切片

以前我们处理的是单个元素的索引。在许多情况下，需要返回多个条目。例如，我们可能希望检索用户6、7和8的记录，而不是第4周用户7的点。我们可以遵循方括号内的切片语法：`start:stop:step`（注意：不包括`stop`索引）：

````python
# note that users 6, 7, and 8 should be indexed by 5, 6, and 7 respectively
# think: why not use 5:7:1 but 5:8:1?
users_678_w4_points = user_info_3d[3, 5:8:1, 2]
print('Points of users 6, 7, and 8 in week 4:', users_678_w4_points)
# When step size is 1, we could skip it
users_678_w4_points = user_info_3d[3, 5:8, 2]
print('Points of users 6, 7, and 8 in week 4:', users_678_w4_points)
````

除了使用整数序列进行索引（即按位置指定元素）之外，我们还可以使用条件来突出显示我们想要的数据类型。假设我们有以下产品销售数据：

````python
rng = np.random.default_rng()
product_sales = rng.integers(low=1, high=50, size=15)
print('Product sales data:', product_sales)
````

假设我们只希望产品销售价值超过20，我们可以：

````python
# we can specify a condition
cond = product_sales > 20
print('Condition is:', cond)
````

与使用整数切片不同，当我们使用布尔值时，方括号中的数组应该与数组大小相同。将返回所有带`True`的元素，而将跳过带`False`的元素：

````python
print('Product sales above 20:', product_sales[cond])
````

我们可能有复合条件，即具有若干标准的条件。例如，我们可能想看看极端值——大于40和小于5的数值:

````python
# we can use the "OR" operatior | to combine conditions together:
# With "|", the resulting boolean value will be True if either condition below is met
cond_extreme = (product_sales>40) | (product_sales<5)
print('Condition is:', cond_extreme)
print('Corresponding value is:', product_sales[cond_extreme])
````

让我们考虑一个更复杂的场景。假设我们的数据有两列，一列表示销售量，另一列表示单价：

````python
rng = np.random.default_rng()
# 10 products
num_products = 10
# 2 attributes: sales and price
num_attr = 2
product_data = rng.integers(low=1, high=50, size=num_products*num_attr).reshape(num_products, num_attr)
print('Product data:\n', product_data)
````

我们如何将此数组子集化，以使结果数据包含以下信息的产品：
- 销售额低于20；
- 价格高于10

````python
# we can specify conditions for seperate columns respectively
# the 1st column is sales
cond_sales = product_data[:, 0] < 20
# the 2nd column is price
cond_price = product_data[:, 1] > 10
# concat the two conditions with `&`
cond = cond_sales & cond_price
print('Products meeting the criteria:\n', product_data[cond])
````

上面要注意的重要一点是冒号`:`的使用，它表示所有元素。更多关于这方面的例子:

````python
# Using `:` in all dimension returns the same values 
print('Return of product_data[:, :]:\n', product_data[:, :])
# First row
print('Return of product_data[0, :]:\n', product_data[0, :])
# Second column
print('Return of product_data[:, 1]:\n', product_data[:, 1])
````

重要的是，当我们使用一个新变量来存储切片产生的值时，我们不会得到一个新元素。相反，我们实际上是在使用这个新变量来指向原始变量的子集部分。例如：

````python
# let's first obtain a subset of `simple_arr`: 1st and 3rd elements
simple_arr = np.array([1, 3, 7, 10, 20, 55])
simple_arr_subset = simple_arr[1:6:2]
print('Before updating the value, simple_arr_subset is:', simple_arr_subset)
print('And simple_arr is:', simple_arr)
simple_arr_subset[0] = 99
print('After updating:', simple_arr_subset)
print('And simple_arr is now:', simple_arr)
````

另一个例子:

````python
print('Before updating the first value, simple_arr_subset is:', simple_arr_subset)
simple_arr_subset[0] *= 2
print('After updating:', simple_arr_subset)
print('And simple_arr is now:', simple_arr)
````

这一点特别重要——当我们对数组进行切片并对其进行计算时，我们必须非常小心。如果我们不打算更改原始对象中的值，我们需要明确表示我们正在复制它，这样新创建的变量只共享该值，而不共享内存地址:

````python
simple_arr = np.array([1, 3, 7, 10, 20, 55])
# This time we explicitly say we want to copy it
simple_arr_subset = simple_arr[1:6:2].copy()
print('Before updating the value, simple_arr_subset is:', simple_arr_subset)
print('And simple_arr is:', simple_arr)
simple_arr_subset[0] = 99
print('After updating:', simple_arr_subset)
print('And simple_arr is now:', simple_arr)
````

但是，值得注意的是，当应用高级索引(相对于基本索引)时，Numpy将返回 _copy_ （深拷贝）而不是 _view_ （浅拷贝）。更具体地说，Numpy定义了:
- 基本切片使用`start:stop:step`来选择给定数组的一个子集。
- 高级索引使用整数或布尔序列进行选取子集(例如，条件逻辑)。

让我们用一个例子来说明这种差异:

````python
simple_arr = np.array([1, 3, 7, 10, 20, 55])
# basic slicing with start:stop:step
subset_basic = simple_arr[0:5:2]
print('Subset with basic slicing:', subset_basic)
# now let's change the value in the subset 
subset_basic[0] = 99
print('subset_basic now:', subset_basic)
print('simple_arr now:', simple_arr)
````

````python
# still, let's use the same array
simple_arr = np.array([1, 3, 7, 10, 20, 55])
# advanced slicing with integer sequence
subset_advanced = simple_arr[np.arange(start=0, stop=5, step=2)]
print('Subset with advanced indexing:', subset_advanced)
# now let's change the value in the subset 
subset_advanced[0] = 99
print('subset_advanced now:', subset_advanced)
print('simple_arr now:', simple_arr)
````

总之，鉴于这种情况，一个建议是运行测试，以检查由于切片(或编码中的任何其他错误)是否有任何意外结果。此外，就我个人而言，在我自己的数据处理中，我几乎总是使用`copy`（深拷贝）方法来避免对原始数组的更改(除非我非常清楚需要同时更新原始数组，这是很少见的)。

### 2.2.3 ndarray的运算

Numpy最有用的特性是用于数字目的。我们从可能经常使用的基本和常见操作开始：

（1）一维数组

我们从一个一维数组开始:假设我们有每个产品被浏览的次数的数据:

````python
rng = np.random.default_rng()
# let's use Poisson distribution for this toy data
# `lam=20` means that the mean is 20
num_prods = 1000
page_views = rng.poisson(lam=20, size=num_prods)
# What does `:10` mean? Run the code and see it yourself!
print('Page view data for the %i products we have:'%num_prods, page_views[:10], '...')
````

在分析这些数据以调查客户活动时，我们经常查看汇总统计数据:

````python
# mean value
page_views_mean = page_views.mean()
# standard deviation (dispersion)
page_views_sd = page_views.std()
print('Mean and SD of page views are: %.3f±%.3f'%(page_views_mean, page_views_sd))
````

极值和总值也是常见被调用的：

````python
page_views_max = page_views.max()
page_views_min = page_views.min()
page_views_sum = page_views.sum()
print('Total page views: %i, with max being %i and min being %i'%(page_views_sum, page_views_max, page_views_min))
````

我们还可以对页面视图数据进行排序，当我们想计算累积和以量化您的产品中是否存在__long tail__。有两种方法对数组进行排序：
- `np.sort`返回已排序数组的副本
- `ndarray.sort`对给定数组进行原地排序(直接更改原始数组中的数据)。

````python
# To avoid messing with our previous generated page view data,
# let's create a small array, which also makes the comparison easier
simple_arr = rng.integers(low=1, high=10, size=5)
print('simple_arr:', simple_arr)
# use np.sort
print('Try running np.sort')
simple_arr_sorted = np.sort(simple_arr)
print('simple_arr:', simple_arr)
print('simple_arr_sorted:', simple_arr_sorted)
# use ndarray.sort
print('Now let\'s try ndarray.sort')
simple_arr_sorted = simple_arr.sort()
print('simple_arr:', simple_arr)
print('simple_arr_sorted:', simple_arr_sorted)
````

有时，我们希望获得数组排序下标的顺序：

````python
simple_arr = np.array([1, 55, 3, 20, 7, 5])
# argsort is the reverse of sort
simple_arr_order = simple_arr.argsort()
print('The order of the elements in simple_arr:', simple_arr_order)
````

我们可以通过索引原始数组来验证此顺序是否确实正确：

````python
print('Indexing simple_arr by the order:', simple_arr[simple_arr_order])
````

（2）二维数组

我们前面所做的一切都可以很容易地应用到n维数组中。但是，值得注意的是，我们需要特别注意将操作应用到哪个维度上。我们将用二维数组来说明这些。

但是，这只能选择第一个或最后几个。当我们有非常具体的条件时，我们可以使用`query`，它实际上等同于布尔索引（即，使用带方括号的条件进行选择），但更灵活，并为代码提供更好的可读性。现在，让我们从一个示例开始，该示例再现了`count `列中超过2行的选择：

````python
# selection made easy by query
print("Selecting rows with at least 3 counts with query:")
print("df.query('count > 2') -->")
print(df.query('count > 2'))
````

````python
# What if we have multiple criteria:
# price is above 18 AND
# count is above 1
# Enter your code below
q = "price>18 and count>1"
print(df.query(q))
````

如果我们的条件涉及变量而不是字面量，会怎么样？我们可以用`@` 表示：

````python
min_count = 1
# the boolean indexing approach
print("Subsetting with boolean indexing: df[df['count']>min_count]")
print(df[df['count']>min_count])
# the query approach
print('Subsetting with query: df.query("count>@min_count")')
print(df.query("count>@min_count"))
````

参数说明如下表所示：

|参数名称|说明|
|:---:|:---|
|<div style="width: 100pt">labels|<div style="width: 400pt">接受 string 或 array，代表删除的行或列的标签，无默认值|

（1）数据清洗——【1】

在数据清洗过程中，主要处理的是缺失值、异常值和重复值。所谓清洗，是对数据集通过丢弃、填充、替换、去重等操作，达到去除异常、纠正错误、补足缺失的目的。

**缺失值的处理**

数据缺失分为两种:一种是行记录的缺失，这种情况又称数据记录丢失;另一种是数据列值的缺失，即由于各种原因导致的数据记录中某些列的值空缺。不同的数据存储和环境中对于缺失值的表示结果也不同，例如，数据库中是 Null，Python 返回对象是 None，Pandas 或 Numpy 中是NaN。

在极少数情况下，部分缺失值也会使用空字符串来代替，但空字符串绝对不同于缺失值。从对象的实体来看，空字符串是有实体的，实体为字符串类型；而缺失值其实是没有实体的，即没有数据类型。

丢失的数据记录通常无法找回，这里重点讨论数据列类型缺失值的处理思路。通常有4种思路。

* 丢弃：

这种方法简单明了，直接删除带有缺失值的行记录（整行删除）或者列字段（整列删除），减少缺失数据记录对总体数据的影响。但丢弃意味着会消减数据特征，以下任何一种场景都不宜采用该方法：

    数据集总体中存在大量的数据记录不完整情况且比例较大，例如超过 10%，删除这些带有缺失值的记录意味着会损失过多有用信息。

    带有缺失值的数据记录大量存在着明显的数据分布规律或特征，例如带有缺失值的数据记录的目标标签(即分类中的 Label 变量)主要集中于某一类或几类，如果删除这些数据记录将使对应分类的数据样本丢失大量特征信息，导致模型过拟合或分类不准确。
    
* 补全

相对丢弃而言，补全是更加常用的缺失值处理方式。通过一定的方法将缺失的数据补上，从而形成完整的数据记录，对于后续的数据处理、分析和建模至关重要。常用的补全方法如下

    统计法：对于数值型的数据，使用均值、加权均值、中位数等方法补足；对于分类型数据，使用类别众数最多的值补足。
    
    模型法：更多时候我们会基于已有的其他字段，将缺失字段作为目标变量进行预测，从而得到最为可能的补全值。如果带有缺失值的列是数值变量，采用回归模型补全；如果是分类变量，则采用分类模型补全。
    
    专家补全：对于少量且具有重要意义的数据记录，专家补足也是非常重要的一种途径。
    
    其他方法：例如随机法、特殊值法、多重填补等
    
* 真值转换

在某些情况下，我们可能无法得知缺失值的分布规律，并且无法对于缺失值采用上述任何一种补全方法做处理；或者我们认为数据缺失也是一种规律，不应该轻易对缺失值随意处理，那么还有一种缺失值处理思路——真值转换。

该思路的根本观点是，我们承认缺失值的存在，并且把数据缺失也作为数据分布规律的一部分，将变量的实际值和缺失值都作为输入维度参与后续数据处理和模型计算中。但是变量的实际值可以作为变量值参与模型计算，而缺失值通常无法参与运算，因此需要对缺失值进行真值转换。

以用户性别字段为例，很多数据库集都无法对会员的性别进行补足，但又舍不得将其丢弃掉，那么我们将选择将其中的值，包括男、女、未知从一个变量的多个值分布状态转换为多个变量的真值分布状态。

    转换前：性别(值域：男、女、未知)。
    转换后:性别_男(值域1或0)、性别_女(值域1或0)、性别_未知(值域1或0)。

然后将这3列新的字段作为输入维度替换原来的1个字段参与后续模型计算。

* 不处理

在数据预处理阶段，对于具有缺失值的数据记录不做任何处理，也是一种思路。这种思路主要看后期的数据分析和建模应用，很多模型对于缺失值有容忍度或灵活的处理方法，因此在预处理阶段可以不做处理。常见的能够自动处理缺失值的模型包括：KNN、决策树和随机森林、神经网络和朴素贝叶斯、DBSCAN(基于密度的带有噪声的空间聚类)等。这些模型对于缺失值的处理思路是：

    忽略，缺失值不参与距离计算，例如 KNN。
    将缺失值作为分布的一种状态，并参与到建模过程，例如各种决策树及其变体。
    不基于距离做计算，因此基于值的距离做计算本身的影响就消除了，例如 DBSCAN。

在数据建模前的数据归约阶段，有一种归约的思路是降维，降维中有一种直接选择特征的方法。假如我们通过一定方法确定带有缺失值(无论缺少字段的值缺失数量有多少)的字段对于模型的影响非常小，那么我们根本就不需要对缺失值进行处理。因此，后期建模时的字段或特征的重要性判断也是决定是否处理字段缺失值的重要参考因素之一。

对于缺失值的处理思路是先通过一定方法找到缺失值，接着分析缺失值在整体样本中的分布占比，以及缺失值是否具有显著的无规律分布特征，然后考虑后续要使用的模型中是否能满足缺失值的自动处理，最后决定采用哪种缺失值处理方法。在选择处理方法时，注意投入的时间、精力和产出价值，毕竟，处理缺失值只是整个数据工作的冰山一角而已。

在数据采集时，可在采集端针对各个字段设置一个默认值。以 MySOL 为例，在设计数据库表时，可通过 default 指定每个字段的默认值，该值必须是常数。在这种情况下，假如原本数据采集时没有采集到数据，字段的值应该为 Null，虽然由于在建立库表时设置了默认值会导致“缺失值”看起来非常正常，但本质上还是缺失的。对于这类数据需要尤其注意。

**异常值的处理**

异常数据是数据分布的常态，处于特定分布区域或范围之外的数据通常会被定义为异常或“噪音”。产生数据“噪音”的原因很多，例如业务运营操作、数据采集问题、数据同步问题等。对异常数据进行处理前，需要先辨别出到底哪些是真正的数据异常。从数据异常的状态看分为两种:

    一种是“伪异常”，这些异常是由于业务特定运营动作产生的，其实是正常反映业务状态，而不是数据本身的异常规律。
    一种是“真异常”，这些异常并不是由于特定的业务动作引起的，而是客观地反映了数据本身分布异常的分布个案。
    
大多数数据挖掘或数据工作中，异常值都会在数据的预处理过程中被认为是噪音而剔除，以避免其对总体数据评估和分析挖掘的影响。但在以下几种情况下，我们无须对异常值做抛弃处理。

* 异常值正常反映了业务运营结果

该场景是由业务部门的特定动作导致的数据分布异常，如果抛弃异常值将导致无法正确反馈业务结果。

例如:公司的 A 商品正常情况下日销量为1000台左右。由于昨日举行优惠促销活动导致总销量达到10000台，由于后端库存备货不足导致今日销量又下降到100台。在这种情况下10000台和100台都正确地反映了业务运营的结果，而非数据异常案例。

* 异常检测模型

异常检测模型是针对整体样本中的异常数据进行分析和挖掘，以便找到其中的异常个案和规律，这种数据应用围绕异常值展开，因此异常值不能做抛弃处理。

异常检测模型常用于客户异常识别、信用卡欺诈、贷款审批识别、药物变异识别、恶劣气象预测、网络人侵检测、流量作弊检测等。在这种情况下，异常数据本身是目标数据，如果被处理掉将损失关键信息。

* 包容异常值的数据建模

如果数据算法和模型对异常值不敏感，那么即使不处理异常值也不会对模型本身造成负面影响。例如在决策树中，异常值本身就可以作为一种分裂节点。

**重复值的处理**

数据集中的重复值包括以下两种情况:

    数据值完全相同的多条数据记录。这是最常见的数据重复情况。
    
    数据主体相同但匹配到的唯一属性值不同。这种情况多见于数据仓库中的变化维度表，同一个事实表的主体会匹配同一个属性的多个值。

去重是重复值处理的主要方法，主要目的是保留能显示特征的唯一数据记录。但当遇到以下几种情况时，请慎重(不建议) 执行数据去重。

* 重复的记录用于分析演变规律

以变化维度表为例。例如在商品类别的维度表中，每个商品对应的同1个类别的值应该是唯一的，例如苹果 iPhone7 属于个人电子消费品，这样才能将所有商品分配到唯一类别属性值中。但当所有商品类别的值重构或升级时(大多数情况下随着公司的发展都会这么做)，原有的商品可能被分配了类别中的不同值。如下表所示：

|:---:|:---:|:---:|
|<div style = "width:100pt">苹果iPhone7|<div style = "width:100pt">个人电子消费品|<div style = "width:100pt">手机数码 |

此时，我们在数据中使用 Full join 做跨重构时间点的类别匹配时，会发现苹果 iPhone7 会同时匹配到个人电子消费品和手机数码2条记录。对于这种情况，需要根据具体业务需求处理。

    如果跟业务沟通，两条数据需要做整合，那么需要确定一个整合字段用来涵盖2条记录。其实就是将2条数据再次映射到一个类别主体中。
    
    如果跟业务沟通，需要同时保存两条数据，那么此时不能做任何处理。后续的具体处理根据建模需求而定。
    
* 重复的记录用于样本不均衡处理
    
在开展分类数据建模工作时，样本不均衡是影响分类模型效果的关键因素之一。解决分类法的一种方法是对少数样本类别做简单过采样，通过随机过采样，采取简单复制样本的策略增加少数类样本。经过这种处理方式后，也会在数据记录中产生相同记录的多条数据。此时,们不能对其中的重复值执行去重操作。

* 重复的记录用于检测业务规则问题

对于以分析应用为主的数据集而言，存在重复记录不会直接影响实际运营，毕竟数据集主是用来做分析的。但对于事务型的数据而言，重复数据可能意味着重大运营规则问题，尤其这些重复值出现在与企业经营中与金钱相关的业务场景时，例如:重复的订单、重复的充值、重复的预约项、重复的出库申请等。

这些重复的数据记录通常是由于数据采集、存储、验证和审核机制的不完善等问题导致的，会直接反映到前台生产和运营系统。以重复订单为例，假如前台的提交订单功能不做唯一性约束，那么在一次订单中重复点击提交订单按钮，就会触发多次重复提交订单的申请记录，如果该操作审批通过后，会联动带动运营后端的商品分拣、出库、送货，如果用户接收重复商品则会导致重大损失；如果用户退货则会增加反向订单，并影响物流、配送和仓储相关的各个运营环节，导致运营资源无端消耗、商品损耗增加、仓储物流成本增加等问题。

因此，这些问题必须在前期数据采集和存储时就通过一定机制解决和避免。如果确实产生了此类问题，那么数据工作者或运营工作者可以基于这些重复值来发现规则漏洞，并配合相关部门，最大限度地降低由此而带来的运营风险。

（2）数据处理——【1】

**数据标准化**

数据标准化是一个常用的数据预处理操作，目的是处理不同规模和量纲的数据，使其缩放到相同的数据区间和范围，以减少规模、特征、分布差异等对模型的影响。除了用作模型计算,标准化后的数据还具有直接计算并生成复合指标的意义，是加权指标的必要步骤。

* 实现中心化和正态分布的 Z-Score

Z-Score 标准化是基于原始数据的均值和标准差进行的标准化，假设原转换的数据为 $x$，新数据为$x'$，那么$x'=\frac{x-mean}{std}$，其中 $mean$ 和 $std$ 为$x$ 所在列的均值和标准差。

这种方法适合大多数类型的数据，也是很多工具的默认标准化方法。标准化之后的数据是以0为均值，方差为1的正态分布。但是 Z-Score 方法是一种中心化方法，会改变原有数据的分布结构，不适合对稀疏数据做处理。

* 实现归一化的 Max-Min

Max-Min 标准化方法是对原始数据进行线性变换，假设原转换的数据为 $x$，新数据为 $x'$，那么$x'=\frac{x-min}{max-min}$，其中 $min$ 和 $max$ 为 $x$ 所在列的最小值和最大值。

这种标准化方法的应用非常广泛，得到的数据会完全落入 $[0,1]$ 区间内(Z-Score 则没有类似区间)。这种方法能使数据归一化而落到一定的区间内，同时还能较好地保持原有数据结构。

* 用于稀疏数据的 MaxAbs

最大值绝对值标准化(MaxAbs)即根据最大值的绝对值进行标准化。假设原转换的数据为 $x$，新数据为 $x'$，那么$x'=\frac{x}{|max|}$，其中 $max$ 为 $x$ 所在列的最大值。

MaxAbs方法与Max-Min用法类似，也是将数据落入一定区间，但该方法的数据区间$[-1,1]$。MaxAbs 也具有不破坏原有数据分布结构的特点，因此也可以用于稀疏数据、稀疏的 CSR(Compressed Sparse Row, 行压缩)或 CSC(Compressed Sparse Column, 列压缩)矩阵。

* 针对离群点的 RobustScaler

某些情况下，假如数据集中有离群点，我们可以使用 Z-Score 进行标准化，但是标准化后的数据并不理想，因为异常点的特征往往在标准化之后容易失去离群特征。此时，可以使用 RobustScaler 针对离群点做标准化处理，该方法对数据中心化和数据的缩放鲁棒性有更强的参数控制。

**数据离散化**

所谓离散化，就是把无限空间中有限的个体映射到有限的空间中。数据离散化操作大多是针对连续数据进行的，处理之后的数据值域分布将从连续属性变为离散属性，这种属性一般包含2个或2个以上的值域。离散化处理的必要性如下：

    节约计算资源，提高计算效率。
    
    算法模型(尤其是分类模型)的计算需要。虽然很多模型，例如决策树可以支持输入连续型数据，但是决策树本身会先将连续型数据转化为离散型数据，因此离散化转换是一个必要步骤。

    增强模型的稳定性和准确度。数据离散化之后，处于异常状态的数据不会明显地突出异常特征，而是会被划分为一个子集中的一部分，因此异常数据对模型的影响会大大降低，尤其是基于距离计算的模型(例如 K均值、协同过滤等)效果明显。
    
    特定数据处理和分析的必要步骤，尤其在图像处理方面应用广泛。大多数图像做特征检测(以及其他基于特征的分析)时，都需要先将图像做二值化处理，二值化也是离散化的一种。
    
    模型结果应用和部署的需要。如果原始数据的值域分布过多，或值域划分不符合业务逻辑，那么模型结果将很难被业务理解并应用。

* 针对时间数据的离散化

针对时间数据的离散化主要用于以时间为主要特征的数据集中和粒度转换，离散化处理后将分散的时间特征转换为更高层次的时间特征。

在带有时间的数据集中，时间可能作为行记录的序列，也可能作为列(维度)记录数据特征。常见的针对时间数据的离散化操作有以下两类：
    
    针对一天中的时间离散化。一般是将时间截转换为秒、分钟、小时或上下午。
    针对日粒度以上数据的离散化。一般是将日期转化为周数、周几、月、工作日或休息日、季度、年等。

针对时间数据的离散化可以将细粒度的时间序列数据离散化为粗粒度的3类数据：

    离散化为分类数据，例如上午、下午。
    离散化为顺序数据，例如周一、周二、周三等。
    离散化为数值型数据，例如一年有52个周，周数是数值型数据。

* 针对多值离散数据的离散化

针对多值离散数据的离散化指的是要进行离散化处理的数据本身不是数值型数据，而是分类或顺序数据。

例如，用户收入变量的值原来可能划分为10个区间，根据新的建模需求，只需要划分为4个区间，那么就需要对原来的10个区间进行合并。

多值离散数据要进行离散化还有可能是划分的逻辑有问题，需要重新划分。这种问题通都是由于业务逻辑的变更，导致在原始数据中存在不同历史数据下的不同值域定义。

例如，用户活跃度变量的值，原来分为高价值、中价值和低价值3个类别。根据业务发展的需要，新的用户活跃度变量的值定义为：高价值、中价值、低价值和负价值。此时需要对同类别的数据进行统一规则的离散化处理。

* 针对连续数据的离散化

针对连续数据的离散化是主要的离散化应用，在分类或关联分析中应用尤其广泛。这些算法的结果以类别或属性标识为基础，而非数值标记。

连续数据的离散化结果可以分为两类：一类是将连续数据划分为特定区间的集合，例如$\left\{(0,10],(10,20],(20,50],(50,100]\right\}$；一类是将连续数据划分为特定类，例如类1、类2、类3。

常见实现针对连续数据化离散化的方法如下：

    分位数法：使用四分位、五分位、十分位等分位数进行离散化处理，这种方法简单易行。
    
    距离区间法：可使用等距区间或自定义区间的方式进行离散化，这种操作更加灵活且能满足自定义需求。另外该方法(尤其是等距区间)可以较好地保持数据原有的分布。
    
    频率区间法：将数据按照不同数据的频率分布进行排序，然后按照等频率或指定频率离散化，这种方法会把数据变换成均匀分布。好处是各区间的观察值是相同的，不足是已经改变了原有数据的分布状态。
    
    聚类法：例如使用K均值将样本集分为多个离散化的簇。
    
    卡方：通过使用基于卡方的离散化方法，找出数据的最佳临近区间并合并，形成较大的区间。

* 针对连续数据的离散化

在很多场景下，我们可能需要将变量特征进行二值化操作：每个数据点跟阙值比较。大于阈值设置为某一固定值(例如1)，小于阈值设置为某一固定值(例如0)，然后得到一个只拥有两个值域的二值化数据集。

二值化应用的前提是数据集中所有的属性值所代表的含义相同或类似，例如读取图像所获得的数据集是颜色值的集合(具体颜色模式取决于读取图像时的模式设置，例如灰度、RGB等)，因此每一个数据点都代表颜色，此时可对整体数据集做二值化处理。某些情况下，也可能只针对特定列做二值化，这样不同列的属性虽然不同，但同一列内产生的二值化结果却仍然具有比较和分类意义。

## 2.4 数据交互

### 2.4.1 数据文件

虽然电子表格很方便，但如果要管理的数据十分复杂，它就不再适用了。事实上，公司更倾向于用数据库（DBs）来维护各种数据，而不是用大量Excel文件。因此，我们有必要了解如何利用Python等工具来访问数据库，在许多情况下，与纯SQL解决方案相比，它的效率更高。

一个数据库由三个部分组成。
- 表: 存储数据的容器。每张表都可以进一步细分为：
    - 行/记录：记录是我们想要保存的项目的独立单位（例如，教师/学生/实习生）
    - 列/属性 ：每条记录的属性（例如，姓名、教职员工ID，等等）
- 元数据：指对存储在表中的数据的描述。虽然这初看起来微不足道，但在实际应用中却很重要，因为在实际应用中，可能需要处理数百个列。
- 关系：大多数情况，我们会使用关系型数据库。这指的是单个表以某种方式连接的事实。比如，有两个表格：（i）产品评论；（ii）产品销售。每个产品都有客户评价和销售量的记录。虽然它们通常存储在不同的地方（例如，客户评价通常是客户关系管理系统[CRM]的一部分，而销售量是组织的内部数据），但这两张表可以根据产品的唯一标识符（例如，通用产品代码[UPC]）进行连接。换句话说，表/数据集之间存在着关系。

数据库中有一个重要概念：_键_ 。键在以下两方面是至关重要的：
- 一个键可以唯一地标识表中的某一行。每张表都必须有一个键（否则我们怎么能定位特定的行呢？）仔细想想--我们能用人名作为包含学生信息的表的键吗？
- 两张表之间的关系基于它们键的重叠。按照前面产品评论和产品销售的例子，我们可以把这两张表的键都定义为UPC。假如有一个应用场景是寻找客户评价高于4.5分的产品的平均价格。首先，我们可以查询数据集子集，只保留评分最低为4.5分的产品。然后我们可以找出这个产品子集的价格并计算其平均值。
	


有两种类型的键：
- _主键_ 是在表中唯一识别行的某列（或多列）。
- _外键_ 是一个来自另一张表中的主键。

请注意，将出现如下报错：
```python
ValueError: Table 'audience_reviews' already exists.
```

这由此函数中参数的默认值导致：
```
if_exists: {‘fail’, ‘replace’, ‘append’}, default ‘fail’
How to behave if the table already exists.
- fail: Raise a ValueError.
- replace: Drop the table before inserting new values.
- append: Insert new values to the existing table.
```

为了解决这个问题，我们只需要特殊说明 `if_exists='append'`:

````python
# try again but say that we want to append instead of making a new table
new_audience_reviews.to_sql(name='audience_reviews', index=False, con=conn, if_exists='append')
````
