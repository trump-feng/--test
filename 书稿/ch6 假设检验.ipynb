# 第六章 假设检验（商品的促销手段）【2+数理统计基础PPT】

参数估计和假设检验是统计推断的两个组成部分，它们都是利用样本对总体进行某种推断，但推断的角度不同。参数估计讨论的是用样本统计量估计总体参数的方法，总体参数在估计前是未知的。而在假设检验中，则是先对从的值提出一个假设，然后利用样本信息去检验这个假设是否成立。因此可以说，本章讨论的内容是如何利用样本信息，对假设成立与否作出判断的一套程序。

## 6.1 假设检验的基本原理

### 6.1.1 假设检验概念

#### （1） 假设检验基本定义

假设检验是对总体参数的具体数值所做的陈述。例如：总体均值大于0、两个总体均值相等、两个变量无关等，都是对总体参数做出的陈述。其基本原理是：逻辑上运用反证法，统计上则依据小概率原理. 即，给定小概率，即在一次试验中，一个几乎不可能发生的事件发生的概率，如果在一次试验中小概率事件一旦发生，我们就有理由拒绝原假设。

#### （2）原假设和备择假设

在统计学中，一般通过等式或不等式来表述假设问题。**原假设**又称为“0假设”，研究者想收集证据予以反对的假设，用$H_0$表示。原假设最初被假设是成立的，之后根据样本数据确定是否有足够的证据拒绝它。常见的原假设有：参数没有变化、或变量之间没有关系、或总体分布与某种理论分布无差异。原假设总是包含符号$=$，具体类型如下：<p>
$$H_0:\mu=\mu_0$$
$$H_0:\mu \geq \mu_0$$
$$H_0:\mu \leq \mu_0$$<p>
**备择假设**也称为“研究假设”，一般是研究者想收集证据予以支持的假设，用$H_1$或$H_\alpha$表示。备择假设所表达的含义总是指参数有变化、或变量之间有关系、或总体分布与某种理论分布有差异等。其通常用于表达研究者自己倾向于支持的看法，然后就是想办法收集证据拒绝原假设，以支持备择假设。备择假设总是包含符号$\neq$、$<$、$>$，具体如下：<p>
$$H_0:\mu \neq \mu_0$$
$$H_0:\mu < \mu_0$$
$$H_0:\mu > \mu_0$$

#### （3） 双边检验与单边检验

备择假设没有特定的方向性，并含有符号$\neq$的假设检验，称为双侧检验或双尾检验。备择假设具有特定的方向性，并含有符号>或<的假设检验，称为单侧检验或单尾检验。另外，当备择假设的方向为<时，称为左侧检验；备择假设的方向为>，称为右侧检验。以总体局长你hi的假设检验为例：<p>


#### （4） 两类错误与显著性水平

对于原假设提出的命题，我们需要作出判断，这种判断可以用“原假设正确”或“原假设错误”来表述。当然，这是依据样本提供的信息进行判断的，也就是由部分来推断总体。因而判断有可能正确，也有可能不正确，也就是说，我们面临着犯错误的可能。所犯的错误有两种类型，一种是弃真错误，另一种是取伪错误<p>
**弃真错误**：也叫第I类错误或$\alpha$错误。它是指原假设实际上是真的，但通过样本估计总体后，拒绝了原假设。明显这是错误的，我们拒绝了真实的原假设，所以叫弃真错误，这个错误的概率我们记为$\alpha$。这个值也是显著性水平，在假设检验之前我们会规定这个概率的大小。<p>
**取伪错误**:也叫第II类错误或$\beta$错误：它是指 原假设实际上假的，但通过样本估计总体后，接受了原假设。明显是错误的，我们接受的原假设实际上是假的，所以叫取伪错误，这个错误的概率我们记为β。

| 项目       | 没有拒绝$H_0$      | 拒绝$H_0$      |
| ---------- | ---------- | ---------- |
| $H_0$为真 | $1-\alpha$(正确决策) | $\alpha$(齐真错误）|
| $H_0$为伪 | $\beta$(取伪错误) |$1-\beta$(正确决策) |

人们希望犯这两类错误的概率越小越好。但对于一定的样本量$n$，不能同时做到犯这两类错误的概率都很小。如果减少$\alpha$错误，就会增大犯$\beta$错误的机会;若减少$\beta$错误，就会增大犯$\alpha$错误的机会。这就像在区间估计中，要想增大估计的可靠性，就会使区间变宽而降低精度;要想提高精度，就要求估计区间变得很窄，而这样，估计的可靠性就会大打折扣。当然，使$\alpha$和$\beta$同时变小的办法也有，就是增大样本量。但样本量不可能没有限制，否则就会使抽样调查失去意义。因此，在假设检验中，就有一个对两类错误进行控制的问题。一般来说，对于一个给定的样本，如果犯第I类错误的代价高于第II类错误，那么将第I类错误的概率定得低些是
合理的；反正，则定得高些一般来说，发生哪一类错误的后果更为严重，就应该首先控制哪一类错误发生的概率，但由于犯第I类错误的概率可以由研究者控制，因此在假设检验中，人们往往先控制第I类错误的发生概率。

**显著性水平**是指事先确定的用于拒绝原假设$H_0$时所必须的证据；能够容忍的犯第I类错误的最大概率(上限值)；原假设为真时，拒绝原假设的概率，即抽样分布的拒绝域。一般用$\alpha$表示，常用的值有0.01、0.05、0.1，由研究者事先确定。

#### （5）检验统计量
根据样本观测结果计算出对原假设和备择假设做出决策的某个样本统计量。对样本统计量的标准化结果：原假设为H0为真、点估计量的抽样分布。<p>
$$标准化检验统计量=\frac{点估计量-假设值}{点估计量的标准误}$$<p>
根据假设检验的不同内容和进行检验的不同条件，需要采用不同的检验统计量，在一个总体参数的检验中，用到的检验统计量主要有三个:$z$统计量，$t$统计量，$\chi_2$统计量。$z$统计量和$t$统计量常常用于均值和比例的检验，$\chi_2$统计量则用于方差的检验。选择什么统计量进行检验需要考虑一些因素，这些因素主要有样本量$n$的大小以及总体标准差是否已知。

### 6.1.2 P值的决策规则

若假设为：$H_0：\mu=500，H_1：\mu<500$。样本均值为495，拒绝$H_0$吗？样本均值为502，拒绝$H_0$吗？做出拒绝或不拒绝的依据是什么？
通常，做出决策的依据是样本的统计量，现代统计检验中，人们直接使用由统计量计算得到的犯第I类错误的概率，即所谓的$P$值.
给定显著水平$\alpha$，查表得出相应的临界值$z_\alpha$或$z_{\alpha/2}$,$t_\alpha$或$t_{\alpha/2}$。将检验统计量的值与α水平的临界值进行比较，做出决策：<p>
    <center>双侧检验：|统计量|>|临界值|，拒绝$H_0$</center><p>
        <center>左侧检验：统计量<-|临界值|，拒绝$H_0$</center><p>
            <center>右侧检验：统计量>|临界值|，拒绝$H_0$</center><p>


### 6.1.3 假设检验流程

假设检验的一般流程如下：<br>
第一步，确定研究的目的，比如说要检验两组数据间的差异，考察某一假设的真伪或是检验某一关系的存在性。<br>
第二步，确定假设检验的类型，比如是双尾检验还是单尾检验，这取决于要测试的假设是双向和无方向性，还是单向和有方向性。<br>
第三步，选择合适的统计检验方法，根据检验类型，选择合适的检验类型，比如 t-检验 、比例检验、卡方检验、F 检验等等。<br>
第四步，计算检验的显著性程度，即计算出概率值(P值)以及统计量，并衡量它们之间的关系，以判断假设的真伪。<br>
第五步，决定采纳或拒绝某一假设，根据计算出的概率值，此时可以采取明确的决定,如果P值小于预定的临界值，则拒绝原假设，也就是说原假设不成立，反之则成立。

### 6.1.4 注意事项
**要证明原假设不正确，P值要多小才能令人信服？**原假设的可信度有多高？如果$H_0$所表示的假设是人们多年来一直相信的，就需要很强的证据(较小的P值)才能说服人们。拒绝的结论是什么？如果拒绝$H_0$而肯定$H_1$，你需要很强的证据支持$H_1$。例如，$H_1$代表要花很多钱把产品包装改成另外一种包装，你就要有很强的证据显示新包装一定会增加销售量(因为拒绝$H_0$要花很高的成本)。<p>
**固定显著水平是否有意义？**若计算出P值，就不需给定显著水平α来做决策。P值为我们决策提供了更多的信息，让我们可以选择任意显著水平来评估结果是否有统计上的显著性，从而根据我们的实际需求决定是否拒绝原假设。传统的显著水平，如0.01、0.05、0.1, 已经被人们普遍接受为"决绝原假设有足够证据"的标准，其中0.1表示有一些证据不利于原假设，0.05表示有适度证据不利于原假设，0.01表示有很强的证据不利于原假设。<p>
**检验假设是否能证明原假设正确？**当不能拒绝原假设时，我们也不能说“接受原假设”。此时，只能说没有足够的证据拒绝原假设，但也没有证据支持原假设，因此无法证明原假设为真。换句话说：没有足够的证据拒绝原假设并不等于你已经"证明"了原假设为真，它仅仅意味着目前还没有找到足够的证据拒绝原假设，只是表示手头的样本提供的证据不足以拒绝原假设。"不拒绝"原假设的表述，实际上仅仅意味着没有得到明确的结论。<p>
**统计学上显著一定有实际意义吗？**当拒绝原假设时，我们称样本结果是统计上显著的；当不拒绝原假设时，我们称样本结果是统计上不显著的；在显著和不显著之间没有清楚的界限，只是在P值越来越小时，我们就有越来越强的证据，检验的结果也就越来越显著。在进行决策时，我们只能说P值越小，拒绝原假设的证据越强，检验结果也就越显著。但P值越小而拒绝原假设时，并不一定意味着检验的结果有意义，必须要结合实际情况来判断。因为P值与样本的大小密切相关，样本量越大，检验统计量的P值也就会越大；若样本量较大时，P值较小，则拒绝原假设的可能性越大。



## 6.2 一个总体参数的假设检验



### 6.2.1 总体均值的假设检验

#### （1）大样本

假设前提：大样本($n\geq 30$)，使用z检验统计量<p>
$\sigma^2$已知，定义检验统计量:<p>
$$z=\frac{\bar x - \mu_0}{\sigma/\sqrt{n}}\sim N(0,1)$$
$\sigma^2$未知，用$s^2$代替，定义检验统计量<p>
$$z=\frac{\bar x - \mu_0}{s/\sqrt{n}}\sim N(0,1)$$

#### （2）小样本

假设前提：（1）总体服从正态分布;（2）小样本($n< 30$)。使用z检验统计量：<p>
$\sigma^2$已知，定义检验统计量:
$$z=\frac{\bar x - \mu_0}{\sigma/\sqrt{n}}\sim N(0,1)$$
$\sigma^2$未知，用$s^2$代替，定义检验统计量：
$$t=\frac{\bar x - \mu_0}{s/\sqrt{n}}\sim t(n-1)$$

<font color='blue'>**【例6.2.1-1：均值，大样本，双侧检验】**</font>：某机床厂加工一种零件，根据经验知道，该厂加工零件的椭圆度渐近服从正态分布，其总体均值为0.081mm，今另换一种新机床进行加工，取200各零件进行检验，得到椭圆度均值为0.076mm，样本标准差为0.025mm，问新机床加工零件的椭圆度总体均值与以前有无差别？

第一步：提出假设：$H_0 : \mu = 0.081mm；H_1 : \mu \neq 0.081mm$,该问题为双边检验，

```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats

# 由题可知
n = 200
mu_0 = 0.081
x_bar = 0.076
sigma = 0.025
a = 0.05

# 计算z值
z_a2 = stats.norm.isf(a/2)
print(z_a2)

# 计算统计量
z = (x_bar-mu_0)/(sigma/np.sqrt(n))
print(z)

# 依据统计量 z 决策
if abs(z)<abs(z_a2):
    print('不能拒绝H0，差异不显著')
if abs(z)>abs(z_a2):
    print('拒绝H0，接受H1')

# 计算p值
p_value = stats.norm.cdf(z)*2
print(p_value)

# 依据p值决策
if p_value > a :
    print('不能拒绝H0，差异不显著')
    
if p_value < a :
    print('拒绝H0，接受H1')
```

<font color='blue'>**例题解析**</font>

通过统计量可以看出，其统计量落在了拒绝域中，则拒绝原假设，接受备择假设，即总体均值与之前有差别。第二种检验方法：P值小于0.05，拒绝原假设。



### 6.2.2 总体方差的假设检验

（1）检验一个总体的方差或标准差，假设总体服从正态分布，使用$\chi^2$分布，检验统计量为：<p>
$$\chi^2 = \frac{(n-1)s^2}{\sigma_0^2}\sim \chi^2(n-1)$$<P>
其中$\sigma_0^2$是假设的总体方差。

$\chi_2$检验分布图

<font color='blue'>**【例6.2.2-1:总体方差】**</font>:某厂商生产出一种新型的饮料装瓶机器，按设计要求，该机器装一瓶1000ml的音量误差上下不超过1ml。如果达到设计要求，表明机器的稳定性非常好。现从该机器装完的产品中随机抽取25瓶，分别进行测定（用样本观测值分别减1000ml），得到如下：0.3,-0.3,-1.3,-0.6,-0.5,-0.4,-1.5,0.7,0.7,1,-0.7,0.6,1,-1.5,-0.2,1.4,-0.9,-0.5,-0.2,-0.6,-0.6,1.3,0,-1.9,1.1。现以a=0.05的显著性水平检验两个总体的方差是否相等。

**第一步：提出原假设和备择假设：**$H_0:\sigma^2 \leq 1 H_1:\sigma^2 >1$,该检验为单边检验，python代码如下：

```python
# 数据导入
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats
data = pd.Series([0.3,-0.3,-1.3,-0.6,-0.5,-0.4,-1.5,0.7,0.7,1,-0.7,0.6,1,-1.5,-0.2,1.4,-0.9,-0.5,-0.2,-0.6,-0.6,1.3,0,-1.9,1.1])


# 由题可得出
n = len(data)
s_2 = data.var()
sigma_2 = 1
a = 0.05

# 计算卡方值
df = n-1
X2_a = stats.chi2.isf(a,df)
print(X2_a)

# 计算统计量
X2 = (n-1)*s_2/sigma_2
print(X2)


# 依据统计量决策
if abs(X2)<abs(X2_a):
    print('不能拒绝H0，差异不显著')
if abs(X2)>abs(X2_a):
    print('拒绝H0，接受H1')
    
# 计算P值
p_value = stats.chi2.sf(X2,df)
print(p_value)

# 依据p值决策
if p_value > a :
    print('不能拒绝H0，差异不显著')
if p_value < a :
    print('拒绝H0，接受H1')
```

<font color='blue'>**例题解析**</font>

通过统计量和P值都可以看出，不能拒绝原假设，即差异不显著。



### 6.2.3 总体比例的假设检验

假设前提：(1)总体服从二项分布；(2)可用正态分布来近似(大样本)。使用z检验统计量：<p>
$$z=\frac{p-\pi_0}{\sqrt{\frac{\pi_0(1-\pi_0)}{n}}} \sim N(0,1)$$
其中$\pi_0$为总体比例

## 6.3 两个总体参数的假设检验



### 6.3.1 两个总体均值之差的假设检验

**（1）大样本**

**假设前提**：(1)两个样本是独立的随机样本；(2)正态总体或非正态总体，大样本($n_1 \geq 30$和$n_2 \geq 30$)。使用z检验统计量<p>
$\sigma_1^2$和$\sigma_2^2$已知，定义检验统计量:<p>
$$z=\frac{(\bar x_1-\bar x_2) - (\mu_1-\mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}}\sim N(0,1)$$<p>
$\sigma_1^2$和$\sigma_2^2$未知，定义检验统计量:<p>
$$z=\frac{(\bar x_1-\bar x_2) - (\mu_1-\mu_2)}{\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}}}\sim N(0,1)$$

**（2）小样本**

**假设前提**：(1)两个样本是独立的随机样本；(2)正态总体或非正态总体，大样本($n_1 < 30$和$n_2 < 30$)；(3) $\sigma_1^2$和$\sigma_2^2$已知。使用z检验统计量：<p>
$\sigma_1^2$和$\sigma_2^2$已知，定义检验统计量:<p>
$$z=\frac{(\bar x_1-\bar x_2) - (\mu_1-\mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}}\sim N(0,1)$$

**假设前提**：(1)两个总体都是正态总体；(2)小样本($n_1 < 30$和$n_2 < 30$)；(3) $\sigma_1^2=\sigma_2^2$未知。使用z检验统计量：<P>
$$t=\frac{(\bar x_1-\bar x_2) - (\mu_1-\mu_2)}{s_p\sqrt{\frac{1}{n_1}+\frac{1}{n_2}}} \sim t(n_1+n_2-2)$$<p>
其中$s_p^2 = \frac{(n_1-1)s_1^2+(n_2-1)s_2^2}{n_1+n_2-2}$

**假设前提**：(1)两个总体都是正态总体；(2)小样本($n_1 < 30$和$n_2 < 30$)；(3) $\sigma_1^2 \neq \sigma_2^2$未知。使用t检验统计量：<p>
$$t=\frac{(\bar x_1-\bar x_2) - (\mu_1-\mu_2)}{\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}}} \sim t(v)$$<p>
其中$v = \frac{(\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2})^2}{
	\frac{(\frac{s_1^2}{n_1})^2}{(n_1-1)}
	+\frac{(\frac{s_2^2}{n_2})^2}{(n_2-1)}}$

**<font color='blue'>【例6.3.1-1：总体均值、方差相同】**</font>:装配一个部件可以采用不同的方法，所关心的问题是哪一个方法的效率更高。劳动效率可以用平均装配时间来反映。现从不同的装配方法中各抽取12件产品，记录各自的装配时间，如下：A:31,34,29,32,35,38,34,30,29,32,31,26；B:26,24,28,29,30,29,32,26,31,29,32,28。两总体为正态总体，且方差相同，问着两种方法的装配时间有无显著差别（a=0.05）。

**第一步：提出原假设和备择假设**：$H_0:\mu_A - \mu_B = 0;H_1:\mu_A - \mu_B \neq 0$，该检验为双侧检验，python代码如下：

```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats

data = pd.DataFrame({'A':[31,34,29,32,35,38,34,30,29,32,31,26],'B':[26,24,28,29,30,29,32,26,31,29,32,28]})

# 由题可得
x1_bar = data.A.mean()
x2_bar = data.B.mean()
s1_2 = data.A.var()
s2_2 = data.B.var()
n1 = len(data.A)
n2 = len(data.B)
a = 0.05

# 计算
 
df = n1+n2-2
t_a2 = stats.t.isf(a/2,df)
print(t_a2)

s_p = np.sqrt(((n1-1)*s1_2+(n2-1)*s2_2)/(n1+n2-2))
print(s_p)


# 计算统计量
t = (x1_bar-x2_bar)/(s_p*np.sqrt(1/n1+1/n2))
print(t)

# 依据统计量t决策

if abs(t)<abs(t_a2):
    print('不能拒绝H0，差异不显著')
if abs(t)>abs(t_a2):
    print('拒绝H0，接受H1')

# 计算p值

p_value = stats.t.sf(t,df)*2
print(p_value)


# 依据p值决策

if p_value > a :
    print('不能拒绝H0，差异不显著')
if p_value < a :
    print('拒绝H0，接受H1')
```

### 6.3.2 两个总体方差的假设检验

**假设前提**：(1)两个总体都服从正态分布，且方差相等；(2)两个独立的随机样本。检验的统计量为： 
$$F = \frac{s_1^2}{s_2^2}\sim F(n_1-1,n_2-1)$$<p>
或
$$F = \frac{s_2^2}{s_1^2}\sim F(n_2-1,n_1-1)$$

<font color='blue'>**【例6.3.2-1：两个总体方差】：**</font>尽管存在争议，但大多数科学家认为，使用含有高纤维的谷类食物有助于降低癌症发生的可能性。然而一个科学家提出，如果人们在早餐中食用高纤维的谷类食物，那么平均而言，与早餐没有食用谷物的人群相比，食用谷物这在午餐中摄取的热量将会减少。如果这个观点成立，谷物食品的生产商又将获得一个很好的机会，他们会宣传说：“多吃谷物吧，早上也吃，这样有助于减肥。”为了检验这个假设，随机抽取了35人，询问他们早餐和午餐的通常食谱，根据他们的食谱将其分为两类，一类为经常谷类食用者（A组），一类为非经常谷类食用者（B组），然后测度没人午餐的大卡摄取量。经过一段时间的实验，得到的结果如下：A组：568,496,589,681,540,646,636,539,596,607,529,617,555,562,584。B组：650,637,563,723,569,628,580,651,622,706,711,569,630,617,480。现以a=0.05的显著性水平检验两个总体的方差是否相等。

**第一步：提出原假设和备择假设**：$H_0:\sigma_1-\sigma_2 =0 $；$H_1：\sigma_1-\sigma_2 \neq 0$,该问题问双边检验。Python计算如下：

```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats


A = pd.Series([568,496,589,681,540,646,636,539,596,607,529,617,555,562,584])
B = pd.Series([650,637,563,723,569,628,580,651,622,706,711,569,630,617,480])

# 由题可得

s1_2 = A.var()
s2_2 = B.var()
n1 = len(A)
n2 = len(B)
alpha = 0.05

# 自由度及F值
df1 = n1-1
df2 = n2-1
f_a2 = stats.f.isf(a/2,df1,df2)
f_1_a2 = stats.f.isf(1-a/2,df1,df2)
F = s1_2/s2_2
print(F)
print('本例中，两个临界点分别为{:.3f}，{:.3f}，F={:.3f}没有落入拒绝域,故不能拒绝H0'.format(f_1_a2,f_a2,F))

# P值
p_value=stats.f.cdf(alpha,df1,df2)
print(p_value)
```

**<font color='blue'>例题解析</font>**

本例中，两个临界点分别为0.336，2.979，F=0.585没有落入拒绝域,故不能拒绝原假设，则A组和B组存在差异；另外从P值中可以看出，p值大于0.05，选择拒绝原假设。

### 6.3.3 两个总体比例的假设检验

**假设前提**：(1)两个总体都服从二项分布；(2)可用正态分布来近似(大样本)。

检验$H_0$：$\pi_1-\pi_2=0$
$$z=\frac{p_1-p_2}{
	\sqrt{ p(1-p) (\frac{1}{n_1}+\frac{1}{n_2}) }
   } \sim N(0,1)$$
其中$p=\frac{x_1+x_2}{n_1+n_2} =\frac{p_1 n_1+p_2 n_2}{n_1+n_2}$<p>
检验$H_0$：$\pi_1-\pi_2=d_0$
$$z=\frac{ (p_1-p_2)-d_0 }{
	\sqrt{(\frac{p_1(1-p_1)}{n_1}+\frac{p_2(1-p_2)}{n_2})} }$$

<font color='blue'>**【例6.3.3-1：检验上网收费比例的差异】**</font>:某公司想研究改进后的上网收费方案满意度是否得到提升。为了验证这个问题，该公司调查了方案修改前后的用户满意度，分别调查300名消费者，其中改进前的方案A有33人满意，改进后的方案B有84人满意。请检改进后的方案B的满意度是否高于方案A。（显著性水平0.05）

**第一步**：根据题设，假设$π_A$、$π_B$为两种方案的满意度比例，因此提出原假设和备择假：$H_0:π_A>π_B$；$H_1:π_A≤π_B$。检验过程如下:<P>

```python
from scipy import stats
import math
n1=300
n2=300
p1=33/300
p2=84/300
p=(p1*n1+p2*n2)/(n1+n2)
z=(p1-p2)/math.sqrt(p*(1-p)*(1/n1+1/n2))
p_value=stats.norm.cdf(z)
print(p_value)
```

**<font color='blue'>例题解析</font>**

p值为7.398020111387353e-08，拒绝原假设，说明改进后的满意度有所提升。

<font color='blue'>**【例6.3.3-2：次品率】**</font>:假设有两种生产方法生产同一种产品，方法1的生产成本高但次品率低，方法2的生产成本高但次品率高。管理人员在选择生产方法时，决定对两种方法的次品率进行比较，如方法1比方法2的次品率低8%以上，则决定采用方法1，否则就采用方法2.管理人员从方法1和方法2生产的产品中，分别随机抽取300个样本，并发现方法1和方法2生产的产品中分别有33、38个次品。（显著性水平0.05）

根据题设，假设$π_1$,$π_2$为两种方案的满意度比例，因此提出原假设和备择假：$H_0:π_1-π_2 \geq 8\%$；$H_1：π_1-π_2 \leq 8\%$。检验过程如下:<P>

```python
from scipy import stats
import math
n1=300
n2=300
p1=33/300
p2=84/300
d0=0.08
z=((p1-p2)-0.08)/math.sqrt(p1*(1-p1)/n1+p2*(1-p2)/n2)
p_value=stats.norm.cdf(z)
print(p_value)
```

**<font color='blue'>例题解析</font>**

p值为1.2634796781451086e-15，拒绝原假设，两种生产力存在差异。

## 6.4 非参数检验

参数检验(如$t$检验，$F$检验等)通常都是假定总体服从正态分布或总体分布形式已知的条件下进行的，而且要求所分析的数据是数值型的。当总体的概率分布形式未知，或者无法对总体的概率分布做出假定时，参数检验方法往往会失效。非参数检验方法不仅对总体的分布要求很少，对数据类型的要求也比参数检验宽松。当数据不适合用参数检验时，非参数检验往往能够得出理想的结果。非参数检验有：<p>
（1）正态性检验：K-S检验<p>
通过数据画出的频数分布的直方图或茎叶图。若数据近似服从正态分布，则图形的形状与上面给出的正态曲线应该相似。一般通过绘制正态概率图来查看数据是否满足正态分布，有时也称为分位数-分位数图，或Q-Q图，或P-P图。Q-Q图则是根据观测值的实际分位数与理论分布的分位数绘制的。P-P图是根据观测数据的累计概率与理论分布(如正态分布)的累积概率的符合程度绘制的。

另外，可以通过Shapiro-Wilk和K-S正态性检验来检验数据是否服从正态性分布。下面着重来介绍下K-S正态性检验。K-S检验是用来检验抽取样本所依赖的总体是否服从某个已知的理论分布。它是将一变量的累积分布函数与特定的分布函数进行比较。设总体的累积分布函数为$F(x)$，已知的理论分布函数$F_0(x)$，则检验的原假设和备择假设分别为:<p>
$$H_0:F(x)=F_0(x)$$
$$H_1:F(x)\neq F_0(x)$$ 

设各样本观测值在理论分布中出现的累积概率为$F_0(x)$，各样本观测值的实际累积概率为$S(x)$，实际累积概率与理论累积概率的差值为$D(x)$，差值序列中的最大绝对差值为：<p>
$$D=max(|S(x_i)-F(x_i)|)$$<p>
由于实际累积概率为离散值，通常作为以下修正<p>
$$D=max\{|S(x_i)-F(x_i)|,|S(x_{i-1})-F(x_i)|\}$$<p>
该式子就是K-S检验的统计量。在小样本情形下，统计量D服从Kolmogorov分布；在大样本情形下，则用正态分布近似，统计量为：$z=\sqrt{n}D$

如果$H_0$成立，则每次抽样得到的$D$值应该不会与0偏离太远，否则就应该拒绝$H_0$。对于设定的显著性水平$\alpha$，若检验统计量$D$(或$z$)的概率小于$\alpha$，则拒绝$H_0$，表示抽样所依赖的总体与指定的理论分布有显著差异。<p>
K-S检验求样本数据是连续的数值数据，而且要求理论分布已知。比如，检验样本数据是否来自$\mu=100、\sigma^2=100$的正态总体，若总体均值和方差未知，则可以用样本均值和方差来代替。


（2）总体位置参数的检验<P>
只有一个总体时，通常关心总体的某个位置参数(如中位数)是否等于某个假定值，检验方法主要有Wilcoxon符号秩检验；当有两个总体时，通常关心两个总体的位置参数是否相同。对于独立样本，采用Mann-Whiteney检验，对于配对样本，则采用两个配对样本的Wilcoxon符号秩检验。下面首先介绍一下秩和结的概念。<p>
**秩**：秩就是一组数据按照从小到大的顺序排列之后，每一个观测值所在的位置。一般符号$R$来表示，假定一组数据$X_1,X_2,...,X_n$，按照从小到大的顺序排列，$X_i$在所有观测值中排第$R_i$位，那么$X_i$的秩即为$R_i$。$R_i$也是一个统计量，它测度的是数据观测值的相对大小，大多数非参数检验方法正是利用秩的这一性质来排除总体分布未知的障碍的。当然，也有一些非参数方法并不涉及秩的性质。<p>
**结**：很多情况下，数据中会出现相同的观测值，那么对它们进行排序后，这些相同观测值的排名显然是并列的，也就是说它们的秩是相同的，这种情况被称为数据中的"结"。对于结的处理，通常是以它们排序后所处的位置的平均值作为它们共同的秩。当一个数据中结比较多时，某些非参数检验中原假设下检验统计量的分布就会受到影响，从而需要对统计量进行修正(一般情况下，软件会自动修正)。

**a.一个总体中位数的Wilcoxon符号秩检验**<p>
**目的**：检验总体参数(如中位数)是否等于某个假定的值。它是对符号检验的一种改进，弥补符号检验的不足，比单纯的符号检验更准确(与单样本均值t检验或z检验相对应)<p>
**检验的步骤**：
第一步：计算各样本观测值与假定的中位数之间的差值，并取绝对值；<p>
第二步：将差值的绝对值排序，并找出它们的秩；<p>
第三步：计算检验统计量和P值，并做出决策。<p>
**b.两个独立样本的Mann-Whitney检验**<p>
Mann-Whitney检验也称作U检验或Wilcoxon秩和检验。其用于确定两个总体间是否存在差异的一种非参数检验方法(与两个独立样本的t检验或z检验相对应)。Mann-Whitney检验不需要做出总体服从正态分布或方差相等的假设，但要求两个独立随机样本的数据至少是顺序数据。与Wilcoxon符号检验不同，它不是基于相关样本，而是使用两个独立样本。<p>
设$X、Y$是两个连续的总体，其累计分布函数为$F_x$和$F_Y$,从两个总体中分别抽取两个独立样本：<p>
$$(X_1,X_2,...,X_m)和(Y_1,Y_2,...,Y_n)$$<p>
若要检验两个总体是否相同，需要提出原假设和备择假设，即：<p>
$$H_0：M_X = M_Y, H1:M_X \neq MY$$<p>
检验的步骤如下：<p>
第一步：把两组数据混合在一起，得到$m+n=N$个数据，并计算$N$个数据的秩；<p>
第二步：分别求样本$(X_1,X_2,...,X_m)$和样本$(Y_1,Y_2,...,Y_n)$的平均值，得到$\bar{W}_x$和$\bar{W}_Y$，并对平均秩的差距进行比较：若二者相差甚远，意味着一组样本的秩普通偏小，另一组样本的秩普遍大，此时原假设有可能不成立。<p>
第三步：计算样本$(X_1,X_2,...,X_m)$中每个秩大于样本$(Y_1,Y_2,...,Y_n)$的每个秩的个数$U_{YX}$，以及样本$(Y_1,Y_2,...,Y_n)$中每个秩大于样本$(X_1,X_2,...,X_m)$的每个秩的个数$U_{XY}$。<p>
第四步：计算Wilcoxon统计量$W$和Mann-Whitney统计量$U$:首先，分别求出两个样本的秩的和，$W_x$和$W_Y$。若$m<n$，检验统计量$W=W_X$；若$m>n$，检验统计量$W=W_Y$；若$m=n$，检验统计量为第一变量值所在样本组的$W$值。Mann-Whitney统计量定义为：$U=W−k(k+1)/2$，其中$k$为$W$对应样本组的样本数据个数。小样本的情况下，统计量服从Mann-Whitney分布，大样本情况下，近似服从正态分布，统计量为：<p>
$$z=\frac{U-nm/2}{\sqrt{mn(m+n+1)/12}}$$<p>
第五步：根据$P$值做出决策。

**c.两个配对样本Wilcoxon符号秩检验**<p>
**目的**：检验两个总体的分布是否相同，或者说两个总体的中位数是否相同，与两个配对样本的$t$检验相对应。<p>
检验统计量：小样本情况下，统计量服从Wilcoxon符号秩和分布：$W=min(W^+,W^-)$。在大样本情况下，统计量近似服从正态分布，为：<p>
$$z=\frac{W-n(n+1)/4}{\sqrt{n(n+1)(2n+1)/24}}$$<p>
检验步骤：<p>
第一步：计算各数据对的差值$d_i$，并取绝对值，排序后求出秩。<P>
第二步：提出检验假设，假设为：$H_0:M_d=0；H1:M_d\neq0$，其中$M_d$表示差值的中位数。并计算检验统计量$W$或$z$。<P>
第三步：根据P值做出决策。
