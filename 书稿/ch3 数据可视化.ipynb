可视化是指将数据通过图形或图像的方法展示出来的一个过程，它将难以直接显示的数据转化成直观的图形、符号等，帮助人们快速理解。<br>
在可视化出现之前，人们通常只能凭借经验对数据进行整合统计分析，对一个领的非专业人士而言，很难通过大量数据快速、准确地理解数据背后所隐含的现象，而视化解决了这个问题，可视化的出现能帮助非专业人士通过图形或图像等轻松挖掘数据背后的意义。因此，进行可视化的原因可以大致概括成以下 3 点。<br>

1.可视化使数据更直观<br>
大量的抽象数字“堆叠”在一起，很难从中快速挖掘有价值的信息，而将抽象的数字转成图表，可以轻松突出数据的重点。<br>
2.可视化可以降低学习成本<br>
可视化能将数字直观地表现出来，非专业人士也可以快速看懂数字，然后深入挖掘数字背后的意义，这样大大降低了非专业人士的学习成本。例如对于公司的利润表，非专业人士很难看懂，但是如果通过可视化将其表现出来，那么非专业人士可以轻松看出公司是否盈利等信息。<br>
3.可视化便于记忆<br>
人脑对图像的敏感度比对文字的敏感度高，通过图表总结大量、复杂的数据便于人脑的记忆与理解。例如，在演讲中使用图表比纯使用文字更加能让观众印象深刻，更有利于让观众理解演讲者演讲的内容。<br>

### 3.1.3 Matplotlib库

参考[1]Python数据可视化 [美]马里奥·多布勒 蒂姆·高博曼

Matplotlib 是较为流行的 Python 绘图库，常用于数据科学和机器学习的可视化操作中。2003 年，约翰·亨特 (John Hunter) 发布了 Matplotlib，旨在模拟 MATLAB 软件中的命令(Matplotlib 软件也是当时的科学标准)。Matplotlib 引入了一些特性，如 MATLAB的全局样式，以便 MATLAB 用户更容易地过渡到 Matplotlib。<br>
在开始与 Matplotlib 协同工作并构建可视化内容之前，下面首先讨论绘图中所涉及的些概念。<br>

Matplotlib 中的图表包含了一种内嵌于 Python 对象的层次结构,进而生成一类树形结构。具体来说，每个图表均封装于 Figure 对象中，即可视化的顶层容器，其中可包含多个轴向一-基本上，这可视为该顶层容器内的独立图表。<br>
在进一步了解后，还会发现 Python 对象可控制轴向、刻度线、图例、标题、文本框网格和许多其他对象。相应地，全部对象均可定制。<br>

图表中的主要组件涵盖以下内容。<br>
* Figure。Figure 是最外层容器并用作绘制的画布，进而可在其中绘制多个图表Figure 不仅加载 Axes 对象，还可对 Title 进行配置。
* Axes。Axes 是实际的图表或子图表，这取决于是否希望绘制单一或多项可视化。
* 内容。Axes 的子对象中包含了x轴和y轴、spine 和图例。<br>
如果从较高的层次上查看这一设计模式，可以看到该层次结构可构建复杂、可定制的可视化内容。<br>

（3）修改轴的元素<p>
plot中缺少一些东西,比如说:<br>
* 没有X/Y标签，因此除了创造这个图的我们自己，没有人知道两个轴上的数值是什么意思。
* X轴刻度线的文字也有重叠,这使得该图的可读性降低。

我们如何处理这个问题呢？让我们来解决第一个问题：分别用[set_xlabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html)和[set_ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html)给轴添加标签。

````python
# 使用subplots创建fig和ax
fig, ax = plt.subplots()
# 现在将数据添加到画布上
scatter_close = ax.scatter(x=btc['Date'], y=btc['Close'])
# 添加 X/Y 标签
xlab = ax.set_xlabel('Date')
ylab = ax.set_ylabel('BTC-USD Closing Price')
````

现在看起来更清晰了! 但是，我们仍然没有解决X轴刻度文字重叠的问题。到目前为止，我们可以尝试两种方法。
- 旋转刻度线文本
- 使文本变小
这两种方法实际上都是在使用同一个方法[set_xticklabels](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticklabels.html)。<br> 
注意，由于我们要修改现有的标签，我们可以先用[get_xticklabels](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.get_xticklabels.html) 把标签提取出来：<br>

````python
# 使用subplots创建fig和ax
fig, ax = plt.subplots()
# 现在将数据添加到画布上
scatter_close = ax.scatter(x=btc['Date'], y=btc['Close'])
# 添加 X/Y 标签
xlab = ax.set_xlabel('Date')
ylab = ax.set_ylabel('BTC-USD Closing Price')
# 让我们更新一下X轴
xlab_values = ax.get_xticks()
xlab_ticks = ax.get_xticklabels()
# 让我们改变一下刻度线的标签
xlab_values = ax.set_xticks(xlab_values)
# 指定字体大小为8和旋转（度）来倾斜标签
xlab_ticks = ax.set_xticklabels(xlab_ticks,
                                rotation=-45,
                                fontdict={'size': 8})
````

鉴于Y轴显示的是比特币的美元价值，我们可能想添加一个美元符号$，使之看起来更好。类似于我们处理X轴刻度线标签的方式，我们可以按照类似的步骤进行。但是在我们继续之前，让我们看一下刻度线内容中的内容。

````python
# 获得Y轴刻度线
ylab_ticks = ax.get_yticklabels()
print('The tick labels of the y-axis:')
print(ylab_ticks)
````

通过查看由`get_yticklabels`返回的[Text](https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text)对象，我们可以知道每个列表中的每个刻度线标签的三个元素分别是x值、y值和文本。因此，我们可以继续使用[set_text](https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text.set_text)函数来更新文本。

````python
# 对于每个元素
for ylab_t in ylab_ticks:
    # 上传文本
    ylab_t_updated = ylab_t.get_text()
    ylab_t.set_text('$%s'%ylab_t_updated)
print('Updated with dollar signs')
print(ylab_ticks)
````

````python
# 使用subplots创建fig和ax
fig, ax = plt.subplots()
# 现在将数据添加到画布上
scatter_close = ax.scatter(x=btc['Date'], y=btc['Close'])
# 添加 X/Y 标签
xlab = ax.set_xlabel('Date')
ylab = ax.set_ylabel('BTC-USD Closing Price')
# 让我们更新一下X轴
xlab_values = ax.get_xticks()
xlab_ticks = ax.get_xticklabels()
# 让我们改变一下刻度线的标签
xlab_values = ax.set_xticks(xlab_values)
# 指定字体大小为8和旋转（度）来倾斜标签
xlab_ticks = ax.set_xticklabels(xlab_ticks,
                                rotation=-45,
                                fontdict={'size': 8})
# 现在更新Y轴的刻度线
ylab_values = ax.get_yticks()
# 更新Y刻度值和样式
ylab_values = ax.set_yticks(ylab_values)
ylab_ticks = ax.set_yticklabels(ylab_ticks)
````

（4）增加更多数据<p>
在现实世界的应用中，我们可能想比较不同数据组的趋势。在这个例子中，假设我们对比较Bitocin和Ether的趋势感兴趣。我们能做什么呢？Matplotlib确实让我们的工作变得简单。让我们从读入数据开始。<br>

````python
eth = pd.read_csv('Data/ETH-USD.csv', parse_dates=['Date'])
eth['Date'] = eth['Date'].dt.date
print('Ether data:')
print(eth.head(2))
````

让我们把这两个数据帧合二为一。回顾一下，我们可以使用`pd.concat`。

````python
# 让我们添加一列，以便我们以后可以区分这两种加密货币
btc['coin'] = 'BTC'
eth['coin'] = 'ETH'
crypto = pd.concat([btc, eth], sort=True, ignore_index=True)
print('Combined data has %i rows'%crypto.shape[0])
````

我们可以使用`for`循环和Pandas的`groupby`来减少代码的行数。为了简单起见，我们先不改变X/Y轴。

````python
# 使用子图创建图和轴
fig, ax = plt.subplots()
for coin, coin_data in crypto.groupby('coin'):
    # 现在将数据添加到画布上
    scatter_close = ax.scatter(x=coin_data['Date'], y=coin_data['Close'])
    # 添加 X/Y 标签
    xlab = ax.set_xlabel('Date')
    ylab = ax.set_ylabel('Closing Price in USD')
````

（5）改变图层元素<p>
问题出在哪里？让我们跳过X/Y轴上的问题，只看数字本身。我想到了几个问题。<br>
- 虽然我们可以猜测蓝色标记是BTC的价格，而橙色标记是ETH的价格，但它们应该被清楚地标明。因此，我们应该添加一个[legend](https://matplotlib.org/stable/api/legend_api.html#matplotlib.legend.Legend)。<br>
- 两种加密货币之间的标度非常不同。这使得我们很难发现橙色标记的趋势。我们可以为不同的数据集使用不同的轴刻度。<br>

让我们一次解决一个问题。让我们首先添加一个传说。事实证明，这很容易。需要两个步骤。
- 我们需要在`scatter`函数中添加一个额外的输入参数`label`。 否则，图例中就没有什么可显示的。
- 我们需要通过使用`ax.legend()`明确表示我们想要一个图例。

````python
#使用subplot创建图和轴
fig, ax = plt.subplots()
for coin, coin_data in crypto.groupby('coin'):
    # 现在将数据添加到画布上
    # 还可以添加一个标签来生成一个图例
    scatter_close = ax.scatter(x=coin_data['Date'],
                               y=coin_data['Close'],
                               label=coin
                              )
    # 添加 X/Y 标签
    xlab = ax.set_xlabel('Date')
    ylab = ax.set_ylabel('Closing Price in USD')
leg = ax.legend()
````

现在的比例看起来不错。-- 因为BTC的价格比ETH的价格高很多。因此，我们可以利用Matplotlib中的[twin axes](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/two_scales.html)为Y轴创建单独的轴，一个用于比特币，另一个用于ETH。还有两件不简单的事情要补充。
- 我们需要指定颜色，因为我们使用的是不同的 `Axes`，默认颜色是蓝色。
- 另外，请注意我们这次是如何添加图例的：我们将图例放在图层中。([参考这里](https://stackoverflow.com/a/47370214)和[这里](https://matplotlib.org/stable/api/legend_api.html?highlight=bbox_transform))。

````python
# 使用subplot创建图和轴
fig, ax_btc = plt.subplots()
# 为eth创建一个双Y轴
ax_eth = ax_btc.twinx()
# 创建一个轴的列表
ax_list = [ax_btc, ax_eth]
# 指定颜色列表
color_list = ['blue', 'orange']
for i, (coin, coin_data) in enumerate(crypto.groupby('coin')):
    # 现在将数据添加到画布上
    ax = ax_list[i]
    # 使用下划线来表示不重要的变量
    _ = ax.scatter(x=coin_data['Date'],
                   y=coin_data['Close'],
                   color=color_list[i],
                   label=coin)
    # 增加 X/Y labels
    xlab = ax.set_xlabel('Date')
    ylab = ax.set_ylabel('Closing Price of %s in USD'%coin)
leg = fig.legend(loc='upper right', bbox_to_anchor=[1,1],
                 bbox_transform=ax_btc.transAxes)
````

其次我们来回顾统计学的几个重要的概念：
* 总体(population)：包含所研究的全部个体的集合
* 样本(sample或individul)：从总体中抽取的一部分元素的集合
* 样本量(sample size)：构成样本的元素的数目，或称样本的容量
* 参数(parameter)：用来描述总体特征的概括性数字度量
* 统计量(statistic)：用来描述样本特征的概括性数字度量

变量(variable) 是描述所观察个体的某种特征，其特点是从一次观察到下一次观察可能会出现不同的结果。变量观察的结果就是数据(data)。<p>
根据计量尺度的不同，可以将变量分为：<br>
分类变量(categorical variable)，取值为对象的属性、类别、或区间的变量，通常说明事物的品质特征，通常用文字来表述，其结果表现为类别，也称为定性变量或品质变量(qualitative variable)；<br>
数值变量(metric variable)，取值为数字尺度的变量，说明事物或现象的数量特征，通常用数值来表现，也称为定量变量(quantitative variable)。<br>

描述统计是一种探索性分析方法，它主要通过数据的数量特征或概括性度量来揭示数据中存在的规律。

（1）统计量：不含未知参数的样本函数。设样本（$X1$,$X2$,...,$Xn$）是来自总体$X$的一个样本，统计量定义为：$T$=$h$($X1$,$X2$,...,$Xn$),常见的统计量包含mean, sd, var, min, max, median, range和 quantile（分位数）

（2）样本均值<br>
平均数(mean)也称均值，它是一组数据相加后除以数据的个数得到的结果。平均数是度量数据集中趋势的常用统计量，在参数估计和假设检验中经常用到。设一组样本数据为$x_1$,$x_2$,…,$x_n$，样本量(本数据的个数)为，则样本平均数用工表示，计算公式为:$$\bar{X} = \sum_{i=1}^n X_{i}$$

（3）样本方差：<br>
如果考虑每个数据$x_i$与其平均数$\bar{x}$之间的差异，以此作为一组数据离散程度的度量结果就要比全距和四分位距更为全面和准确。这就需要求出每个数据$x_i$与其平均数$\bar{x}$离差的平均数。但由于$(x_i-\bar{x})$之和等于 0，需要进行一定的处理。一种方法是将离差取绝对值，求和后再平均，这一结果称为平均差 (mean deviation) 或称平均绝对离差 (meanabsolute deviation);另一种方法是将离差平方后再求平均数，这一结果称为方差(variance)。
$$S^2=\frac{1}{n-1}\sum_{i=1}^n (x_{i}-\bar{x})^2 $$

（4）样本标准差：<br>
方差开方后的结果称为标准差 (standard deviation)。方差 (或标准差)是实际中应用最广泛的测度数据离散程度的统计量。<br>$$ S=\sqrt{S^2}$$

（5）变异系数：<br>标准差是反映数据离散程度的绝对值，其数值的大小受原始数据取值大小的影响，数据的观测值越大，标准差的值通常也就越大。此外，标准差与原始数据的计量单位相同，采用不同计量单位计量的数据，其标准差的值也就不同。因此，对于不同组别的数据，如果原始数据的观测值相差较大或计量单位不同，就不能用标准差直接比较其离散程度，这时需要计算离散系数。<br>
离散系数 (coefficient of variation，CV) 也称变异系数，它是一组数据的标准差与其相应的平均数之比。由于离散系数消除了数据取值大小和计量单位对标准差的影响，因而可以反映一组数据的相对离散程度。<br>
离散系数主要用于比较不同样本数据的离散程度。离散系数大说明数据的相对离散程度也大，离散系数小说明数据的相对离散程度也小.<br>$$CV=\frac{s}{\bar{X}}\times 100\% $$

变异系数刻画数据相对分散性。当需要比较两组数据离散程度大小的时候，如果两组数据的测量尺度相差太大，或者数据量纲的不同，直接使用标准差来进行比较不合适，此时就应当消除测量尺度和量纲的影响，而变异系数可以做到这一点，他是标准差与其平均数的比；当CV>15%时，说明数据离散程度过高，需要注意。

（6）最小值、最大值和极差: $$min$$ $$max$$ $$range$$

（7）1/4位数、3/4位数：25th and 75th quartiles<br>
一组数据从小到大排序后，找出排在某个位置上的数值，并用该数值代表数据水平的高低，则这些位置上的数值就是相应的分位数 (quantile)，包括中位数、四分位数等。<br>
中位数 (median) 是一组数据排序后处在中间位置上的数值，用$M_e$表示。中位数是用一个点将全部数据等分成两部分，每部分包含 50%的数据，一部分数据比中位数大，另一部分比中位数小。中位数是用中间位置上的值代表数据的水平，其特点是不受极端值的影响，在研究收入分配时很有用。<br>
计算中位数时，要先对 n 个数据从小到大进行排序，然后确定中位数的位置，最后确定中位数的具体数值。如果位置是整数值，中位数就是该位置所对应的数值;如果位置是整数加 0.5的数值，中位数就是该位置两侧值的平均值。<br>
四分位数 (quartile) 是一组数据排序后处于 25%和75%位置上的数值。它是用3个点将全部数据等分为 4 部分，其中每部分包含 25%的数据。很显然，中间的四分位数就是中位数，因此通常所说的四分位数是指处在 25%位置上和处在 75%位置上的两个数值。<br>
与中位数的计算方法类似，计算四分位数时，首先对数据进行排序，然后确定四分位数所在的位置，该位置上的数值就是四分位数。与中位数不同的是，四分位数位置的确定方法有多种，每种方法得到的结果可能会有一定差异，但差异不会很大 (一般相差不会超过一个位次)。由于不同软件的计算方法可能不一样，因此，对同一组数据使用不同软件得到的四分位数结果也可能会有所差异，但不会影响分析的结论。

（8）偏度：<br>
偏度 (skewness)是指数据分布的不对称性，这一概念由统计学家皮尔逊 (K.Pearson)于 1895 年首次提出。测度数据分布不对称性的统计量称为偏度系数( coefficient of skewness)，记为 SK。根据原始数据计算偏度系数时，通常采用下面的公式:
$$skewness=\frac{n}{(n-1)(n-2)s^3} \sum_{i=1}^n (X_{i}-\bar{X})^3$$

当数据对称分布时，偏度系数等于0。偏度系数越接近 0，偏斜程度就越低，就越接近对称分布。如果偏度系数明显不等于0，表示分布是非对称的。若偏度系数大于1或小于-1，视为严重偏斜分布;若偏度系数在 0.5～1或-1～-0.5 之间，视为中等偏斜分布;偏度系数在 0～0.5或-0.5～0之间时，视为轻微偏斜。其中负值表示左偏分布 (在分布的左侧有长尾)，正值则表示右偏分布 (在分布的右侧有长尾)。

（9）峰度：<br>
峰度 (kurtosis) 是指数据分布峰值的高低，这一概念由统计学家皮尔逊于1905 年首次提出。测度一组数据分布峰值高低的统计量称为峰度系数 (coefficient of kurtosis)，记作 K。根据原始数据计算峰度系数时，通常采用下面的公式:
$$kurtosis=\frac{n(n+1)}{(n-1)(n-2)(n-3)s^4} \sum_{i=1}^n (X_{i}-\bar{X})^4-3\frac{(n-1)^2}{(n-2)(n-3)}$$
峰度通常是与标准正态分布相比较而言的。由于标准正态分布的峰度系数为 0，当K>0时为尖峰分布，数据分布的峰值比标准正态分布高，数据相对集中;当K<0时为扁平分布，数据分布的峰值比标准正态分布低，数据相对分散。

## 3.1 单变量的描述统计

#### 3.2.2.2  分组的数值型变量基于图形的描述统计

除了对单个分组的数值变量进行展示，也可以同时对多个分组的数值变量进行展示，通常使用箱线图和直方图等。<p>
箱形图也称箱线图，它不仅可用于反映一组数据分布的特征，比如，分布是否对称等，还可以对多组数据的分布特征进行比较，这也是箱形图的主要用途。绘制箱形图的步骤大致如下:首先，找出一组数据的中位数 (median)和两个四分位数(quartiles)，并画出箱子。中位数是一组数据排序后处在 50%位置上的数值。四分位数是一组数据排序后处在25%位置和75%位置上的两个分位数值，分别用$Q_{25}\%$和$Q_{75}\%$表示。$Q_{25}\%$-$Q_{75}\%$称为四分位差或四分位距 (quartile deviation)，用IQR表示。用两个四分位数画出箱子(四分位差的范围)，并画出中位数在箱子里面的位置。<br>
其次，计算出内围栏和相邻值，并画出须线。内围栏(inter fence) 是与$Q_{25}\%$和$Q_{75}\%$的距离等于1.5倍四分位差的两个点，其中 $Q_{25}\%$-1.5XIQR称为下内围栏，$Q_{75}\%$+1.5XIQR 称为上内围栏。上下内围栏一般不在箱形图中显示，只是作为确定离群点的界限然后找出上下内围栏之间的最大值和最小值 (即非离群点的最大值和最小值)，称为相邻值 (adjacent value)，其中$Q_{25}\%$-1.5XIQR 范围内的最小值称为下相邻值，$Q_{75}\%$+1.5XIQR 范围内的最大值称为上相邻值。用直线将上下相邻值分别与箱子连接，称为须线(whiskers)。<br>
最后，找出离群点，并在图中单独标出。离群点 (outlier) 是大于上内围栏或小于下内围栏的数值，也称外部点 (outside value)，在图中用空心圈单独标出。<br>
箱线图由一组数据的最大值、最小值、中位数、四分位数等五个值构成，也可以用于描述分组数据的分布状态。

### 3.3.1 气泡图——参考大数据可视化技术 姜枫 许桂秋主编

气泡图与散点图相似，其不同之处在于，气泡图允许用户在图表中额外加入一个表示大小的变量。实际上，这就如同以二维方式绘制包含三个变量的图表。气泡由大小不同的标记（指示相对重要程度）表示。这种图表类型的优势在于它便于我们同时比较三个变量，如图3-x所示。一个变量用x轴表现，另一个变量用y轴表现，而第三个变量则通过气泡的面积大小来体现。

读者需要注意圆形大小是与面积相关的，也就是与半径的平方相关。下面通过实例讲述如何创建气泡图，其Python程序如下，最终生成的图表如图3-x所示。

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
crime =pd.read_csv("crimeRatesByState2005.csv")
print(list(crime.murder))
crime2=crime[crime.state !="United States"]
crime2=crime2[crime2.state!="District of Columbia"]
z=list(crime2.population/10000)
colors=np.random.rand(len(list(crime2.murder)))
cm=plt.cm.get_cmap('RdYlBu')
plt.scatter(list(crime2.murder),list(crime2.burglary),s=z,c=z,cmap=cm,linewidth=0.5,alpha=0.5)
plt.xlabel("murder")
plt.ylabel("burglary")
plt.show()
````

图3-x所示还是谋杀案和入室盗窃案的相关数据，同时加入了各州的人口数量作为第三个变量，通过气泡的大小表示。那么，是否人口越多，犯罪率就越高呢？事实上并非如此。像加利福尼亚州、佛罗里达州和得克萨斯州这些大州确实很接近图表的右上部分，但是，纽约州和宾夕法尼亚州的人口数量也很大，但入室盗窃案发生率却相对较低。与之类似，路易斯安那州和马里兰州的人口较少，但却位于图表的最右边，表明其谋杀案发生率很高。

气泡图的设计过程需要注意以下问题。
* 散点图的设计规则同样适用于气泡图
* 考虑到过多的气泡使得图表难以阅读，因而应避免使用过多的数据量。

### 3.3.2 雷达图——参考Python数据可视化 [美]马里奥·多布勒

雷达图也称作蜘蛛图或Web图，并可对多个变量进行可视化。其中，每个变量绘制于自身轴向上，并生成一个多边形。另外，全部轴向以放射状排列，始于中心位置且彼此间等距，同时包含了相同的尺度。

雷达图的应用场景包括以下方面。
* 雷达图非常适合比较单个分组或多个分组的多个定量变量。
* 有助于显示数据集中哪些变量的评分较高或较低。对于可视化性能来说，雷达图可视为一种较为理想的选择方案。

雷达图的设计过程应注意以下问题。
* 尝试在一个雷达图上显示10个或更少的因素，以便更容易阅读
* 对多变量/分组采用分面，以维护应有的清晰度。
